## 1. Decision Guide

```python
# Quick reference:
# - Ordered, mutable, duplicates allowed: list
# - Ordered, immutable, duplicates allowed: tuple
# - Unordered, mutable, unique elements: set
# - Unordered, mutable, key-value mapping: dict
# - Hashable (immutable) elements: frozenset
```

## 2. Performance Characteristics

```python
# Membership test: list vs set
import time

n = 1_000_000
big_list = list(range(n))
big_set = set(big_list)

start = time.perf_counter()
print(999_999 in big_list)
list_time = time.perf_counter() - start

start = time.perf_counter()
print(999_999 in big_set)
set_time = time.perf_counter() - start

print(f"List: {list_time:.4f}s, Set: {set_time:.4f}s")
```

## 3. Examples: Good Choices

```python
# BAD: list for unique visitors
# visitors = ["ip1", "ip2", "ip1"]  # duplicates possible, O(n) check

# GOOD: set for unique visitors
visitors = {"ip1", "ip2"}
visitors.add("ip1")  # no duplication
print(visitors)
```

```python
# BAD: list for student grade lookup
# students = [("Alice", "A"), ("Bob", "B")]  # O(n) lookup

# GOOD: dict for student grade lookup
grades = {"Alice": "A", "Bob": "B"}
print(grades["Alice"])  # O(1)
```

## 4. Finding Duplicates

```python
def find_duplicates(nums):
    seen = set()
    dupes = set()
    for n in nums:
        if n in seen:
            dupes.add(n)
        seen.add(n)
    return dupes

print(find_duplicates([1, 2, 3, 2, 4, 3, 5]))  # {2, 3}
```

## Summary
- **List**: ordered, mutable, duplicates OK — use for sequences
- **Tuple**: ordered, immutable — use for fixed data, dict keys
- **Set**: unordered, unique — use for membership, dedup
- **Dict**: key-value mapping — use for lookups by key
- O(1) for set/dict membership vs O(n) for list
