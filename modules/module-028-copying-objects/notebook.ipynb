# Module 028: Copying Objects (Shallow vs Deep Copy)

Phase 3: Data Structures | 1.5 hours

Understand references, shallow copies, and deep copies.

## 1. Reference vs Copy

```python
original = [1, 2, 3]
ref = original          # reference — same object
copy = original.copy()  # copy — new object

ref[0] = 99
print("original:", original)  # [99, 2, 3]
print("ref:", ref)            # [99, 2, 3]
print("copy:", copy)          # [1, 2, 3]
```

## 2. Shallow Copy — .copy() and copy.copy()

```python
# List .copy()
nums = [10, 20, 30]
c = nums.copy()
c[0] = 99
print(nums)  # [10, 20, 30]
print(c)     # [99, 20, 30]
```

```python
# Shallow copy limitation — inner objects are shared
nested = [[1, 2], [3, 4]]
shallow = nested.copy()
shallow[0][0] = 99
print("nested:", nested)    # [[99, 2], [3, 4]] — changed!
print("shallow:", shallow)  # [[99, 2], [3, 4]]
```

## 3. Deep Copy — copy.deepcopy()

```python
import copy

nested = [[1, 2], [3, 4]]
deep = copy.deepcopy(nested)
deep[0][0] = 99
print("nested:", nested)  # [[1, 2], [3, 4]] — unchanged!
print("deep:", deep)      # [[99, 2], [3, 4]]
```

## 4. Dict Copying

```python
d = {"items": [1, 2, {"nested": True}]}
shallow = d.copy()
deep = copy.deepcopy(d)

shallow["items"][2]["nested"] = False
print("original:", d)      # {'items': [1, 2, {'nested': False}]} — changed

deep["items"][2]["nested"] = "modified"
print("original:", d)      # unchanged
print("deep:", deep)
```

## Summary
- Assignment (`=`) creates a reference, not a copy
- `.copy()` and `copy.copy()` create shallow copies (inner objects are shared)
- `copy.deepcopy()` creates a fully independent copy
- Use deep copy for nested mutable structures
