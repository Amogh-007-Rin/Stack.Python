# Module 023: Dictionaries: Basics

Phase 3: Data Structures | 2 hours

Learn about dictionaries — Python's key-value mapping data structure.

## 1. Creating Dictionaries

```python
# Using curly braces
person = {
    "name": "Alice",
    "age": 30,
    "city": "New York"
}
print(person)
print(type(person))
```

```python
# Empty dict
empty = {}
also_empty = dict()
print(len(empty), len(also_empty))
```

## 2. Accessing Values

```python
# Using [] — raises KeyError if missing
print(person["name"])

# Using .get() — safe, returns None or default
print(person.get("name"))
print(person.get("country", "Unknown"))
```

## 3. Adding and Modifying Entries

```python
person["email"] = "alice@example.com"  # add new
person["age"] = 31                       # update existing
print(person)
```

## 4. Iterating Dictionaries

```python
scores = {"Alice": 85, "Bob": 92, "Charlie": 78}

for name in scores:
    print(name)

for score in scores.values():
    print(score)

for name, score in scores.items():
    print(f"{name}: {score}")
```

## 5. len() and in

```python
print("Number of students:", len(scores))
print("Alice in scores:", "Alice" in scores)
print("David in scores:", "David" in scores)
```

## Summary
- Dictionaries map keys to values using `{}`
- Access with `[]` (risky) or `.get()` (safe)
- Add/update with `d[key] = value`
- Iterate with `.keys()`, `.values()`, `.items()`
- Check membership with `in`
