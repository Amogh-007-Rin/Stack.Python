## Generator Functions with yield

Use `yield` instead of `return` to produce values one at a time:

```python
def count_up_to(n):
    """Yield numbers from 1 to n.

    Args:
        n: Upper limit (inclusive)

    Yields:
        Numbers from 1 to n
    """
    i = 1
    while i <= n:
        yield i
        i += 1

for num in count_up_to(5):
    print(num, end=" ")
```

## Using next() and StopIteration

Call `next()` manually to get values one by one:

```python
gen = count_up_to(3)
print(next(gen))
print(next(gen))
print(next(gen))
try:
    print(next(gen))
except StopIteration:
    print("Generator exhausted!")
```

## Generator Expressions

Like list comprehensions but with parentheses:

```python
squares = (x ** 2 for x in range(10))
print(type(squares))

for s in squares:
    print(s, end=" ")
```

## Memory Efficiency

Generators don't store all values in memory:

```python
import sys

list_comp = [x for x in range(100000)]
gen_exp = (x for x in range(100000))

print(f"List size: {sys.getsizeof(list_comp)} bytes")
print(f"Generator size: {sys.getsizeof(gen_exp)} bytes")
```

## yield from

Delegate to another generator:

```python
def chain(*iterables):
    """Chain multiple iterables into one."""
    for it in iterables:
        yield from it

result = list(chain("abc", [1, 2, 3], (True, False)))
print(result)
```

## Infinite Sequences

Generators can represent infinite data:

```python
def fibonacci():
    """Generate infinite Fibonacci sequence."""
    a, b = 0, 1
    while True:
        yield a
        a, b = b, a + b

fib = fibonacci()
first_10 = [next(fib) for _ in range(10)]
print(first_10)

first_20 = [next(fib) for _ in range(10)]
print(first_20)
```

## Generator vs List Comparison

Generators are single-use; lists are reusable:

```python
nums_list = [x * 2 for x in range(5)]
nums_gen = (x * 2 for x in range(5))

print("List (can iterate multiple times):")
print(list(nums_list))
print(list(nums_list))

print("Generator (single use):")
print(list(nums_gen))
print(list(nums_gen))  # empty!
```

## Key Takeaways

- `yield` produces a value and pauses; resumes on next `next()`.
- Generator expressions: `(x for x in iterable)`.
- Generators are memory-efficient for large/infinite data.
- `yield from` delegates to sub-generators.
- Generators are single-use (exhausted after iteration).
