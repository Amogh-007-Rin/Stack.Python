# Module 053: Dataclasses

```python
# Basic dataclass
from dataclasses import dataclass


@dataclass
class Point:
    x: float
    y: float


p = Point(3.0, 4.0)
print(p)
```

```python
# field() with defaults and factory
from dataclasses import field
from typing import List


@dataclass
class Student:
    name: str
    grades: List[int] = field(default_factory=list)
    active: bool = True


s = Student("Alice")
s.grades.append(95)
print(s)
```

```python
# frozen=True for immutability
@dataclass(frozen=True)
class ImmutablePoint:
    x: float
    y: float


ip = ImmutablePoint(1, 2)
print(ip)
# ip.x = 5  # would raise FrozenInstanceError
```

```python
# Dataclass inheritance
@dataclass
class Animal:
    name: str
    sound: str

    def speak(self) -> str:
        return f"{self.name} says {self.sound}"


@dataclass
class Dog(Animal):
    breed: str = "mixed"


d = Dog("Rex", "Woof", "German Shepherd")
print(d.speak())
print(d)
```
