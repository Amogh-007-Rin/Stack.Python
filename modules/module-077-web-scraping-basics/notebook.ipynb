## Parsing HTML with BeautifulSoup

```python
from bs4 import BeautifulSoup

html = """
<html>
  <head><title>My Page</title></head>
  <body>
    <h1>Welcome</h1>
    <p class="content">Hello, World!</p>
    <a href="https://example.com">Example</a>
  </body>
</html>
"""

soup = BeautifulSoup(html, 'lxml')
print('Title:', soup.title.text)
print('H1:', soup.h1.text)
print('Link:', soup.a.text, '->', soup.a['href'])
```

## Finding Elements: find() and find_all()

```python
from bs4 import BeautifulSoup

html = """
<ul>
  <li class="item active">Apple</li>
  <li class="item">Banana</li>
  <li class="item">Cherry</li>
</ul>
"""

soup = BeautifulSoup(html, 'lxml')

# find first li
first = soup.find('li')
print('First li:', first.text)

# find all li
all_items = soup.find_all('li')
for item in all_items:
    print('Item:', item.text)

# find by class
active = soup.find(class_='active')
print('Active:', active.text)
```

## CSS Selectors with select()

```python
from bs4 import BeautifulSoup

html = """
<div id="content">
  <p class="first">Paragraph 1</p>
  <p>Paragraph 2</p>
  <div class="inner">
    <p>Nested paragraph</p>
  </div>
</div>
"""

soup = BeautifulSoup(html, 'lxml')

# All p tags inside #content
for p in soup.select('#content p'):
    print('Selected:', p.text)

# First paragraph with class
first = soup.select('.first')
print('First class:', first[0].text if first else 'None')
```

## Navigating the Parse Tree

```python
from bs4 import BeautifulSoup

html = """
<div>
  <p>First</p>
  <p>Second</p>
</div>
"""

soup = BeautifulSoup(html, 'lxml')
first_p = soup.find('p')

print('Parent:', first_p.parent.name)
print('Next sibling:', first_p.find_next_sibling('p').text)
print('Children of div:')
for child in soup.find('div').children:
    if child.name:
        print(' ', child.name, '->', child.text)
```
