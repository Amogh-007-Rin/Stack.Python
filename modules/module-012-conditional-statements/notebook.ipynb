# Module 012: Conditional Statements (if / elif / else)

Make decisions in your code with conditional statements.

## 1. Basic `if` Statement

```python
age = 18
if age >= 18:
    print("You are an adult.")
    print("You can vote.")
```

```python
age = 16
if age >= 18:
    print("This won't print.")
print("This prints regardless.")
```

## 2. `if-else`

```python
age = 16
if age >= 18:
    print("You are an adult.")
else:
    print("You are a minor.")
```

## 3. `if-elif-else` Chain

```python
score = 85
if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
elif score >= 60:
    grade = "D"
else:
    grade = "F"
print(f"Score: {score}, Grade: {grade}")
```

## 4. Truthiness in Conditions

You don't need `== True` — just check the value directly.

```python
name = "Alice"
if name:
    print(f"Hello, {name}!")

empty_name = ""
if not empty_name:
    print("Name is empty!")
```

## 5. Checking for `None`

```python
result = None
if result is None:
    print("No result yet.")
else:
    print(f"Result: {result}")

result = 42
if result is not None:
    print(f"Result is: {result}")
```

## 6. Nested Conditionals

```python
age = 25
has_license = True

if age >= 18:
    if has_license:
        print("You can drive.")
    else:
        print("Get a license first.")
else:
    print("Too young to drive.")
```

## 7. Common Patterns

```python
# Check if a number is in range
num = 15
if num > 0 and num <= 100:
    print(f"{num} is between 1 and 100")

# Ternary-style (conditional expression)
x = 10
result = "positive" if x > 0 else "non-positive"
print(result)
```

```python
# Password length checker
password = "secret123"
if len(password) < 8:
    print("Password too short!")
elif len(password) < 12:
    print("Password is okay.")
else:
    print("Strong password!")
```

```python
# Numeric range with and
temp = 22
if temp > 30:
    print("Hot!")
elif temp > 20:
    print("Warm.")
elif temp > 10:
    print("Cool.")
else:
    print("Cold!")
```
