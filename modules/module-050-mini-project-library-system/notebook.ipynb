## Book Class

Encapsulation, dunder methods, and properties.

```python
class Book:
    """Represents a book in the library."""

    def __init__(self, title, author, isbn, copies=1):
        self.title = title
        self.author = author
        self.isbn = isbn
        self._copies = copies
        self._available = copies

    @property
    def available(self):
        return self._available

    @property
    def copies(self):
        return self._copies

    def borrow(self):
        if self._available <= 0:
            return False
        self._available -= 1
        return True

    def return_copy(self):
        if self._available < self._copies:
            self._available += 1
            return True
        return False

    def __str__(self):
        return f"{self.title} by {self.author}"

    def __repr__(self):
        return f"Book({self.title!r}, {self.author!r}, {self.isbn!r}, {self._copies})"

    def __eq__(self, other):
        if not isinstance(other, Book):
            return NotImplemented
        return self.isbn == other.isbn

    def __hash__(self):
        return hash(self.isbn)


b = Book("Python 101", "Alice", "123-456", 3)
print(b)
print(repr(b))
print(f"Available: {b.available}")
b.borrow()
print(f"After borrow: {b.available}")
```

## Member Class Hierarchy

Inheritance, polymorphism, and borrowing limits.

```python
class Member:
    """Base member class."""

    def __init__(self, name, member_id):
        self.name = name
        self.member_id = member_id
        self.borrowed_books = []

    @property
    def borrowing_limit(self):
        return 0

    @property
    def loan_period_days(self):
        return 0

    def can_borrow(self):
        return len(self.borrowed_books) < self.borrowing_limit

    def borrow_book(self, book):
        if not self.can_borrow():
            raise RuntimeError(f"{self.name} has reached the borrowing limit of {self.borrowing_limit}")
        self.borrowed_books.append(book)

    def return_book(self, book):
        if book not in self.borrowed_books:
            raise ValueError(f"{self.name} did not borrow {book.title}")
        self.borrowed_books.remove(book)

    def __str__(self):
        return f"{self.name} (ID: {self.member_id})"

    def __repr__(self):
        return f"{type(self).__name__}({self.name!r}, {self.member_id!r})"


class StudentMember(Member):
    """Student member with lower borrowing limits."""

    @property
    def borrowing_limit(self):
        return 3

    @property
    def loan_period_days(self):
        return 14


class TeacherMember(Member):
    """Teacher member with higher borrowing limits."""

    @property
    def borrowing_limit(self):
        return 10

    @property
    def loan_period_days(self):
        return 30


members = [
    StudentMember("Alice", "S001"),
    TeacherMember("Dr. Bob", "T001"),
]
for m in members:
    print(f"{m}: limit={m.borrowing_limit}, loan={m.loan_period_days}d")
```

## Library Class

Putting it all together.

```python
class Library:
    """Manages books and members."""

    def __init__(self, name):
        self.name = name
        self._books = {}
        self._members = {}

    def add_book(self, book):
        self._books[book.isbn] = book

    def add_member(self, member):
        self._members[member.member_id] = member

    def borrow_book(self, member_id, isbn):
        member = self._members.get(member_id)
        book = self._books.get(isbn)
        if not member:
            raise ValueError(f"No member with ID {member_id}")
        if not book:
            raise ValueError(f"No book with ISBN {isbn}")
        if not book.borrow():
            raise RuntimeError(f"No copies of {book.title} available")
        member.borrow_book(book)

    def return_book(self, member_id, isbn):
        member = self._members.get(member_id)
        book = self._books.get(isbn)
        if not member or not book:
            raise ValueError("Invalid member or book")
        member.return_book(book)
        book.return_copy()

    def list_books(self):
        for book in self._books.values():
            status = f"({book.available}/{book.copies} available)"
            print(f"  {book} {status}")

    def list_members(self):
        for member in self._members.values():
            borrowed = len(member.borrowed_books)
            print(f"  {member} - {borrowed} books borrowed")

    def __str__(self):
        return f"Library: {self.name}"

    def __len__(self):
        return len(self._books)


lib = Library("City Library")
lib.add_book(Book("Python 101", "Alice", "123", 3))
lib.add_book(Book("Advanced Python", "Bob", "456", 2))
lib.add_member(StudentMember("Alice", "S001"))
lib.add_member(TeacherMember("Dr. Bob", "T001"))

print(f"{lib} ({len(lib)} books)")
print("\nBooks:")
lib.list_books()
print("\nMembers:")
lib.list_members()
print("\nBorrowing...")
lib.borrow_book("S001", "123")
lib.list_books()
lib.list_members()
```

## Key Takeaways

- OOP organizes complex systems into interacting classes
- Inheritance models specialized behavior
- Encapsulation protects internal state
- Dunder methods integrate with Python built-ins
- Polymorphism enables flexible, reusable code
