## 1. Chained Comparisons

```python
x = 5
print(f"1 < x < 10: {1 < x < 10}")
print(f"0 <= x <= 5: {0 <= x <= 5}")
print(f"1 < x < 4: {1 < x < 4}")
```

```python
# Chained comparisons with different operators
age = 25
print(f"18 <= age <= 65: {18 <= age <= 65}")

value = 10
print(f"0 < value <= 100: {0 < value <= 100}")
```

## 2. Combining Comparisons with `and` / `or` / `not`

```python
age = 22
has_ticket = True

if age >= 18 and has_ticket:
    print("Welcome to the show.")

if age < 12 or age >= 65:
    print("Discount applies.")
else:
    print("Standard price.")
```

## 3. Short-Circuit Evaluation

```python
def check(n):
    print(f"  check({n}) called")
    return n > 0

print("and chain (stops at first False):")
result = check(5) and check(0) and check(3)
print(f"Result: {result}")
```

```python
print("or chain (stops at first True):")
result = check(0) or check(-1) or check(5) or check(2)
print(f"Result: {result}")
```

## 4. De Morgan's Laws

```python
age = 20
has_id = False

# These two are equivalent:
if not (age >= 18 and has_id):
    print("Cannot enter (De Morgan form 1).")

if (age < 18) or (not has_id):
    print("Cannot enter (De Morgan form 2).")
```

## 5. `is` vs `==`

```python
a = [1, 2, 3]
b = [1, 2, 3]
c = a

print(f"a == b: {a == b}")  # True (same value)
print(f"a is b: {a is b}")  # False (different objects)
print(f"a is c: {a is c}")  # True (same object)
```

```python
# Always use is for None
value = None
print(f"value is None: {value is None}")
print(f"value is not None: {value is not None}")
```

## 6. The `in` Operator

```python
print(f"'a' in 'hello': {'a' in 'hello'}")
print(f"'e' in 'hello': {'e' in 'hello'}")
print(f"3 in [1, 2, 3]: {3 in [1, 2, 3]}")
print(f"10 not in [1, 2, 3]: {10 not in [1, 2, 3]}")
```

```python
# Practical: vowel checker
char = "a"
if char in "aeiou":
    print(f"{char} is a vowel")
else:
    print(f"{char} is a consonant")
```

## 7. Operator Precedence

```python
x = 5
y = 10
z = 0

# Precedence: comparisons first, then and/or
result = x > 0 and y < 20 or z == 0
print(f"x > 0 and y < 20 or z == 0: {result}")

# With parentheses for clarity
result2 = (x > 0 and y < 20) or (z == 0)
print(f"With parentheses: {result2}")
```
