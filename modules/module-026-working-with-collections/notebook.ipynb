## 1. zip() — Combine Iterables

```python
names = ["Alice", "Bob", "Charlie"]
scores = [85, 92, 78]

zipped = zip(names, scores)
print(list(zipped))
# [('Alice', 85), ('Bob', 92), ('Charlie', 78)]
```

```python
# Unzipping with zip(*pairs)
pairs = [("x", 1), ("y", 2), ("z", 3)]
letters, nums = zip(*pairs)
print(letters, nums)
```

## 2. enumerate() — Indexed Iteration

```python
fruits = ["apple", "banana", "cherry"]
for i, fruit in enumerate(fruits):
    print(f"{i}: {fruit}")
```

```python
# Custom start index
for i, fruit in enumerate(fruits, start=1):
    print(f"{i}. {fruit}")
```

## 3. itertools Overview

```python
from itertools import count, cycle, repeat, chain, product, permutations, combinations

# count — infinite arithmetic progression
for i, n in enumerate(count(10, 5)):
    if i >= 5:
        break
    print(n, end=" ")  # 10 15 20 25 30
print()

# cycle — infinite repetition
c = 0
for item in cycle("AB"):
    if c >= 6:
        break
    print(item, end=" ")
    c += 1
print()  # A B A B A B
```

```python
# chain — combine iterables
print(list(chain([1, 2], [3, 4], [5, 6])))
# [1, 2, 3, 4, 5, 6]

# product — Cartesian product
print(list(product("AB", [1, 2])))
# [('A',1), ('A',2), ('B',1), ('B',2)]
```

```python
# permutations — ordered arrangements
print(list(permutations("ABC", 2)))
# [('A','B'), ('A','C'), ('B','A'), ('B','C'), ('C','A'), ('C','B')]

# combinations — unordered selections
print(list(combinations("ABC", 2)))
# [('A','B'), ('A','C'), ('B','C')]
```

## Summary
- `zip()` pairs elements from multiple iterables
- `enumerate()` adds indices during iteration
- `itertools` provides: `count`, `cycle`, `chain`, `product`, `permutations`, `combinations`
