## Step 2: Terminal Basics

In a real terminal, you would use commands like `pwd`, `ls`, and `cd`. Inside this notebook, we can simulate the experience:

```python
import subprocess
print("Current directory:", end=" ")
print(subprocess.run(["pwd"], capture_output=True, text=True).stdout.strip())
print("Files here:", end=" ")
print(subprocess.run(["ls"], capture_output=True, text=True).stdout.strip())
```

## Step 5: Creating a .py File (Simulated)

In a real setup, you would create a `hello.py` file and run `python3 hello.py` in the terminal.
The code below shows what the file would contain:

```python
content = '''print("Hello from hello.py!")
print("This is my first Python script.")'''
with open("/tmp/hello.py", "w") as f:
    f.write(content)
print("Created /tmp/hello.py")
```

```python
import subprocess
result = subprocess.run(["python3", "/tmp/hello.py"], capture_output=True, text=True)
print(result.stdout)
```

## Key Takeaways

- VS Code + Python extension gives you a modern editing experience.
- Run code in the REPL (`python3`) or as a script (`python3 file.py`).
- Always save your file before running it.
