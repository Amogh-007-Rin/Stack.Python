## Connecting and Creating Tables

```python
import sqlite3

conn: sqlite3.Connection = sqlite3.connect('example.db')
cursor: sqlite3.Cursor = conn.cursor()

cursor.execute('''
    CREATE TABLE IF NOT EXISTS users (
        id INTEGER PRIMARY KEY,
        name TEXT NOT NULL,
        email TEXT UNIQUE,
        age INTEGER
    )
''')
conn.commit()
print('Table created successfully')
```

## Inserting Data (Parameterized Queries)

```python
users: list[tuple[str, str, int]] = [
    ('Alice Johnson', 'alice@example.com', 30),
    ('Bob Smith', 'bob@example.com', 25),
    ('Carol Davis', 'carol@example.com', 35),
]

cursor.executemany(
    'INSERT INTO users (name, email, age) VALUES (?, ?, ?)',
    users,
)
conn.commit()
print(f'Inserted {cursor.rowcount} users')
```

## Querying Data (SELECT)

```python
cursor.execute('SELECT * FROM users')
rows: list = cursor.fetchall()
for row in rows:
    print(f'ID: {row[0]}, Name: {row[1]}, Email: {row[2]}, Age: {row[3]}')
```

## Filtering and Parameterized Queries

```python
# Safe parameterized query - prevents SQL injection
min_age: int = 28
cursor.execute('SELECT name, age FROM users WHERE age > ?', (min_age,))
results: list = cursor.fetchall()
print(f'Users older than {min_age}:')
for name, age in results:
    f'  {name} ({age})'
```

## Updating Data

```python
cursor.execute(
    'UPDATE users SET age = ? WHERE name = ?',
    (31, 'Alice Johnson'),
)
conn.commit()
print(f'Updated {cursor.rowcount} row(s)')

# Verify
cursor.execute('SELECT name, age FROM users WHERE name = ?', ('Alice Johnson',))
print(cursor.fetchone())
```

## Deleting Data

```python
cursor.execute('DELETE FROM users WHERE name = ?', ('Bob Smith',))
conn.commit()
print(f'Deleted {cursor.rowcount} row(s)')

# Confirm deletion
cursor.execute('SELECT COUNT(*) FROM users')
print(f'Remaining users: {cursor.fetchone()[0]}')
```

## Cleanup: Close Connection

```python
conn.close()
print('Connection closed')
```
