## Project Overview

You'll build a menu-driven CLI application that manages contacts:

Data is stored in a dict: `{name: {name, phone, email}}`

## Data Structure Design

````python
# Each contact is a dict with name, phone, email
contact = {
    "name": "Alice",
    "phone": "555-1234",
    "email": "alice@example.com"
}

# All contacts stored in a dict keyed by name
contacts = {}
contacts["Alice"] = contact
print(contacts["Alice"]["email"])  # alice@example.com
````
