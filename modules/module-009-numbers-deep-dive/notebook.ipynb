## Integer Precision

Python integers can be arbitrarily large.

```python
small = 42
huge = 10 ** 100
print("Small:", small)
print("Huge:", huge)
print("Type:", type(huge))
```

## Float Precision

Floats have limited precision. Some decimal values cannot be represented exactly.

```python
print("0.1 + 0.2 =", 0.1 + 0.2)
print("0.1 + 0.2 == 0.3:", 0.1 + 0.2 == 0.3)
print("round(0.1 + 0.2, 2):", round(0.1 + 0.2, 2))
```

### Safe Float Comparison

Use a tolerance instead of `==`.

```python
tolerance = 1e-10
result = abs((0.1 + 0.2) - 0.3) < tolerance
print("Within tolerance?", result)
```

## The `round()` Function

```python
print(round(3.14159, 2))   # 3.14
print(round(3.14159))      # 3
print(round(2.5))          # 2 (bankers' rounding)
print(round(3.5))          # 4 (bankers' rounding)
```

## The `math` Module

Import `math` to access mathematical functions and constants.

```python
import math

print("sqrt(16):", math.sqrt(16))
print("pow(2, 10):", math.pow(2, 10))
print("ceil(3.2):", math.ceil(3.2))
print("floor(3.8):", math.floor(3.8))
print("pi:", math.pi)
print("e:", math.e)
print("factorial(5):", math.factorial(5))
print("gcd(12, 8):", math.gcd(12, 8))
```

## Complex Numbers

Python supports complex numbers natively.

```python
z = 3 + 4j
print("z:", z)
print("real:", z.real)
print("imag:", z.imag)
print("z * z:", z * z)
print("abs(z):", abs(z))
```

## Geometry Calculator Program

```python
import math

print("=== Circle ===")
r = 5.0
print(f"Circumference: {2 * math.pi * r:.2f}")
print(f"Area: {math.pi * r ** 2:.2f}")

print("\n=== Right Triangle ===")
a, b = 3.0, 4.0
c = math.sqrt(a ** 2 + b ** 2)
print(f"Hypotenuse: {c}")

print("\n=== Float Precision Demo ===")
print(f"0.1 + 0.2 = {0.1 + 0.2}")
print(f"0.1 + 0.2 == 0.3 is {0.1 + 0.2 == 0.3}")
print(f"Safe compare: {abs((0.1 + 0.2) - 0.3) < 1e-10}")
```

## Key Takeaways

- Integers: arbitrary precision (no overflow).
- Floats: ~15-17 significant digits; some values inexact.
- Use tolerance for float comparison: `abs(a - b) < 1e-10`.
- `round()` uses bankers' rounding (half to even).
- `import math` for `sqrt`, `ceil`, `floor`, `pi`, `e`, `factorial`, `gcd`.
- `math.ceil()` rounds up, `math.floor()` rounds down.
- Complex numbers: `3 + 4j` with `.real` and `.imag`.
