Understand how mutability affects assignment, function calls, and data structure choices.

## 1. Mutable vs Immutable Types

```python
# Immutable: int, float, str, tuple, frozenset
# Mutable: list, dict, set

# Immutable — cannot change in place
s = "hello"
# s[0] = "H"  # TypeError
s = "Hello"  # reassignment creates a new object

# Mutable — can change in place
lst = [1, 2, 3]
lst[0] = 99
print(lst)  # [99, 2, 3]
```

## 2. Mutability and Function Calls

```python
# Immutable argument — not modified
def increment(x):
    x += 1
    print("Inside:", x)

n = 5
increment(n)
print("Outside:", n)  # 5
```

```python
# Mutable argument — can be modified
def add_item(lst, item):
    lst.append(item)

items = [1, 2]
add_item(items, 3)
print(items)  # [1, 2, 3]
```

## 3. Dict Keys and Set Elements

```python
# Only hashable (immutable) types can be dict keys
d = {}
d[("a", "b")] = "works"  # tuple — OK

try:
    d[["a", "b"]] = "fails"  # list — TypeError
except TypeError as e:
    print(f"Error: {e}")
```

```python
# Same restriction applies to set elements
s = set()
s.add((1, 2))  # OK
try:
    s.add([1, 2])  # TypeError
except TypeError as e:
    print(f"Error: {e}")
```

## 4. Reference Aliasing

```python
a = [1, 2, 3]
b = a  # b is a reference to the same list
b[0] = 99
print("a:", a)  # [99, 2, 3] — a changed too!
print("b:", b)  # [99, 2, 3]
print("Same object?", a is b)  # True
```

## Summary
- Immutable types cannot be changed in place; modifications create new objects
- Mutable types can be modified in place
- Only immutable (hashable) types can be dict keys or set members
- Assignment with `=` creates a reference, not a copy, for mutable objects
- Function arguments: mutable objects can be modified inside the function
