## 1. Data Design

We'll use two parallel lists to store tasks and their completion status.

```python
tasks = []
completed = []

# Simulate adding some tasks
tasks.append("Buy milk")
completed.append(False)
tasks.append("Read book")
completed.append(False)
tasks.append("Write code")
completed.append(True)

for i in range(len(tasks)):
    status = "[✓]" if completed[i] else "[ ]"
    print(f"{i + 1}. {status} {tasks[i]}")
```

## 2. Displaying Tasks

```python
def show_tasks(tasks, completed):
    if not tasks:
        print("No tasks yet!")
        return
    for i in range(len(tasks)):
        status = "[✓]" if completed[i] else "[ ]"
        print(f"{i + 1}. {status} {tasks[i]}")

show_tasks(tasks, completed)
```

## 3. Adding Tasks

```python
def add_task(tasks, completed, description):
    if not description:
        print("Task cannot be empty!")
        return False
    tasks.append(description)
    completed.append(False)
    print(f"Added: {description}")
    return True

add_task(tasks, completed, "Walk the dog")
show_tasks(tasks, completed)
```

## 4. Marking Tasks Complete

```python
def mark_complete(tasks, completed, index):
    if 0 <= index < len(tasks):
        if completed[index]:
            print("Task already complete!")
            return False
        completed[index] = True
        print(f"Completed: {tasks[index]}")
        return True
    else:
        print("Invalid index.")
        return False

mark_complete(tasks, completed, 0)
show_tasks(tasks, completed)
```

## 5. Deleting Tasks

```python
def delete_task(tasks, completed, index):
    if 0 <= index < len(tasks):
        removed = tasks.pop(index)
        completed.pop(index)
        print(f"Deleted: {removed}")
        return True
    else:
        print("Invalid index.")
        return False

delete_task(tasks, completed, 2)
show_tasks(tasks, completed)
```

## 6. Menu Loop Structure

```python
def run_menu():
    local_tasks = []
    local_done = []

    while True:
        print("\n===== TO-DO LIST =====")
        print("1. View tasks")
        print("2. Add task")
        print("3. Mark complete")
        print("4. Delete task")
        print("5. Quit")
        choice = input("Choose (1-5): ")

        if choice == "1":
            show_tasks(local_tasks, local_done)
        elif choice == "2":
            desc = input("Enter task: ")
            add_task(local_tasks, local_done, desc)
        elif choice == "3":
            show_tasks(local_tasks, local_done)
            if local_tasks:
                try:
                    idx = int(input("Complete task #: ")) - 1
                    mark_complete(local_tasks, local_done, idx)
                except ValueError:
                    print("Enter a number.")
        elif choice == "4":
            show_tasks(local_tasks, local_done)
            if local_tasks:
                try:
                    idx = int(input("Delete task #: ")) - 1
                    delete_task(local_tasks, local_done, idx)
                except ValueError:
                    print("Enter a number.")
        elif choice == "5":
            print("Goodbye!")
            break
        else:
            print("Invalid. Enter 1-5.")

# Uncomment to run the interactive menu:
# run_menu()
print("Menu ready — call run_menu() to start.")
```
