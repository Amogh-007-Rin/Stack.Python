# Module 017: Nested Loops and Pattern Printing

Combine loops within loops to create visual patterns.

## 1. Basic Nested Loops

```python
for i in range(3):
    for j in range(4):
        print(f"({i},{j})", end=" ")
    print()
```

## 2. Rectangle Pattern

```python
rows = 3
cols = 7
for i in range(rows):
    for j in range(cols):
        print("*", end="")
    print()
```

## 3. Right Triangle

```python
n = 5
for i in range(1, n + 1):
    for j in range(i):
        print("*", end="")
    print()
```

## 4. Inverted Right Triangle

```python
n = 5
for i in range(n, 0, -1):
    for j in range(i):
        print("*", end="")
    print()
```

## 5. Pyramid

```python
n = 5
for i in range(1, n + 1):
    for j in range(n - i):
        print(" ", end="")
    for j in range(2 * i - 1):
        print("*", end="")
    print()
```

## 6. Diamond

```python
n = 5
for i in range(1, n + 1):
    for j in range(n - i):
        print(" ", end="")
    for j in range(2 * i - 1):
        print("*", end="")
    print()
for i in range(n - 1, 0, -1):
    for j in range(n - i):
        print(" ", end="")
    for j in range(2 * i - 1):
        print("*", end="")
    print()
```

## 7. Multiplication Table

```python
for i in range(1, 10):
    for j in range(1, 10):
        print(f"{i*j:3}", end=" ")
    print()
```

## 8. Number Pyramid

```python
n = 5
for i in range(1, n + 1):
    for j in range(n - i):
        print(" ", end="")
    for j in range(1, i + 1):
        print(j, end="")
    for j in range(i - 1, 0, -1):
        print(j, end="")
    print()
```

## 9. Checkerboard

```python
size = 4
for i in range(size):
    for j in range(size):
        if (i + j) % 2 == 0:
            print("■", end=" ")
        else:
            print("□", end=" ")
    print()
```

## 10. O(n²) Demonstration

```python
n = 500
ops = 0
for i in range(n):
    for j in range(n):
        ops = ops + 1
print(f"n={n}: {ops} operations (n² = {n*n})")
```
