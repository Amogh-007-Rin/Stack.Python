# Module 069: Logging

```python
import logging

# basicConfig: quick setup
logging.basicConfig(level=logging.DEBUG, format="%(levelname)s: %(message)s")

logging.debug("Debug message")
logging.info("Info message")
logging.warning("Warning message")
logging.error("Error message")
logging.critical("Critical message")
```

```python
# Logging to a file
logging.basicConfig(
    level=logging.DEBUG,
    format="%(asctime)s - %(levelname)s - %(message)s",
    filename="app.log",
    filemode="w",
    force=True,
)
logging.info("Application started")
logging.warning("Disk space low")
logging.error("Connection failed")

with open("app.log") as f:
    print(f.read())
```

```python
# Logger objects with hierarchy
logger = logging.getLogger("app")
logger.setLevel(logging.DEBUG)

handler = logging.StreamHandler()
handler.setFormatter(logging.Formatter("%(name)s - %(levelname)s: %(message)s"))
logger.addHandler(handler)

logger.info("App logger info")

# Child logger inherits parent settings
child = logging.getLogger("app.network")
child.warning("Network warning")
```

```python
# Multiple handlers with different formatters
multi_logger = logging.getLogger("multi")
multi_logger.setLevel(logging.DEBUG)

# Console handler: only WARNING and above
console = logging.StreamHandler()
console.setLevel(logging.WARNING)
console.setFormatter(logging.Formatter("[CONSOLE] %(levelname)s: %(message)s"))

# File handler: all levels
file_h = logging.FileHandler("multi.log", mode="w")
file_h.setLevel(logging.DEBUG)
file_h.setFormatter(logging.Formatter("%(asctime)s - %(levelname)s - %(message)s"))

multi_logger.addHandler(console)
multi_logger.addHandler(file_h)

multi_logger.debug("Debug (file only)")
multi_logger.info("Info (file only)")
multi_logger.warning("Warning (console + file)")
```

```python
# Custom formatter
custom_fmt = logging.Formatter(
    "[%(asctime)s] [%(name)s] %(levelname)s: %(message)s",
    datefmt="%Y-%m-%d %H:%M:%S",
)
handler = logging.StreamHandler()
handler.setFormatter(custom_fmt)

log = logging.getLogger("custom_formatter")
log.addHandler(handler)
log.setLevel(logging.INFO)
log.info("This message uses a custom formatter")
```

```python
import os
for f in ["app.log", "multi.log"]:
    try:
        os.remove(f)
    except FileNotFoundError:
        pass
print("Cleanup done")
```
