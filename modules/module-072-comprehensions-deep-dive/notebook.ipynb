## List Comprehensions

```python
# Basic list comprehension
squares = [x**2 for x in range(10)]
print('squares:', squares)

# With condition
even_squares = [x**2 for x in range(10) if x % 2 == 0]
print('even squares:', even_squares)

# With if-else
labels = ['even' if x % 2 == 0 else 'odd' for x in range(5)]
print('labels:', labels)
```

## Dict Comprehensions

```python
# Dict comprehension
square_dict = {x: x**2 for x in range(5)}
print('square dict:', square_dict)

# With condition
even_square_dict = {x: x**2 for x in range(10) if x % 2 == 0}
print('even square dict:', even_square_dict)
```

## Set Comprehensions

```python
# Set comprehension
numbers = [1, 2, 2, 3, 3, 3, 4]
unique_squares = {x**2 for x in numbers}
print('unique squares:', unique_squares)
```

## Generator Expressions

```python
# Generator expression
gen = (x**2 for x in range(5))
print('type:', type(gen))
print('list from gen:', list(gen))

# Memory efficient sum
total = sum(x**2 for x in range(1000000))
print('sum of squares:', total)
```

## Nested Comprehensions

```python
# Flatten a matrix
matrix = [[1, 2], [3, 4], [5, 6]]
flat = [item for row in matrix for item in row]
print('flattened:', flat)

# Nested dict comprehension
table = {i: {j: i*j for j in range(1, 4)} for i in range(1, 4)}
print('multiplication table:')
for k, v in table.items():
    print(f'  {k}: {v}')
```
