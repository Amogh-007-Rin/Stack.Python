# Module 015: Loops: `for`

Iterate over sequences with the `for` loop.

## 1. Basic `for` Loop

```python
for name in ["Alice", "Bob", "Charlie"]:
    print(f"Hello, {name}!")
```

## 2. Iterating Over Strings

```python
word = "Python"
for letter in word:
    print(letter, end=" ")
print()

# Count specific letters
count = 0
for ch in "hello world":
    if ch == 'l':
        count = count + 1
print(f"'l' count: {count}")
```

## 3. `range()` Function

```python
print("range(5):", end=" ")
for i in range(5):
    print(i, end=" ")
print()

print("range(2, 7):", end=" ")
for i in range(2, 7):
    print(i, end=" ")
print()

print("range(0, 10, 2):", end=" ")
for i in range(0, 10, 2):
    print(i, end=" ")
print()

print("range(10, 0, -2):", end=" ")
for i in range(10, 0, -2):
    print(i, end=" ")
print()
```

## 4. Multiplication Table

```python
n = 7
print(f"{n} Times Table:")
for i in range(1, 11):
    print(f"{n} x {i:2} = {n * i:2}")
```

## 5. `for-else` Clause

```python
numbers = [1, 3, 5, 7, 9]
for n in numbers:
    if n % 2 == 0:
        print(f"Found even: {n}")
        break
else:
    print("No even numbers found.")
```

## 6. Iterating with Index

```python
word = "hello"
for i in range(len(word)):
    print(f"Index {i}: {word[i]}")
```

## 7. Practical: Vowel Counter

```python
text = "Hello World"
vowels = "aeiou"
count = 0
for char in text.lower():
    if char in vowels:
        count = count + 1
print(f"Vowels in '{text}': {count}")
```

## 8. Sum with `range()`

```python
total = 0
for i in range(1, 101):
    total = total + i
print(f"Sum 1-100: {total}")
```
