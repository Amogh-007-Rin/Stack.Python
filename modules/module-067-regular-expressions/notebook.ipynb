# Module 067: Regular Expressions

```python
import re

# re.search - find first match anywhere
text = "My email is alice@example.com and backup is bob@test.org"
match = re.search(r"\w+@\w+\.\w+", text)
if match:
    print(f"Found: {match.group()} at position {match.start()}")
```

```python
# re.match - match only at the beginning
print(re.match(r"\d+", "123abc"))  # Match
print(re.match(r"\d+", "abc123"))  # None
```

```python
# re.findall - find all matches
emails = re.findall(r"\w+@\w+\.\w+", text)
print("All emails:", emails)
```

```python
# re.sub - substitute matches
censored = re.sub(r"\w+@\w+\.\w+", "[REDACTED]", text)
print(censored)
```

```python
# re.split - split on pattern
parts = re.split(r"[,;\s]+", "apple, banana; cherry   date")
print(parts)
```

```python
# Metacharacters and character classes
examples = [
    (r"\d+", "Price: 42 dollars"),
    (r"\w+", "hello_world"),
    (r"\s+", "word1   word2"),
    (r"^Start", "Start of string"),
    (r"end$", "This is the end"),
    (r"colou?r", "color colour"),
    (r"\bcat\b", "The cat categorizes"),
]
for pattern, test_str in examples:
    result = re.findall(pattern, test_str)
    print(f"r'{pattern:12s}' in '{test_str:30s}' -> {result}")
```

```python
# Groups and named groups
log_line = "2025-06-15 14:30:00 ERROR Connection timeout from 192.168.1.1"
pattern = r"(?P<date>\d{4}-\d{2}-\d{2})\s+(?P<time>\d{2}:\d{2}:\d{2})\s+(?P<level>\w+)\s+(?P<message>.+)"
m = re.match(pattern, log_line)
if m:
    print(f"Date:    {m.group('date')}")
    print(f"Time:    {m.group('time')}")
    print(f"Level:   {m.group('level')}")
    print(f"Message: {m.group('message')}")
```

```python
# Compiling patterns with re.compile
phone_pattern = re.compile(r"\((\d{3})\)\s\d{3}-\d{4}")
texts = ["Call (555) 123-4567", "Reach (555) 987-6543"]
for t in texts:
    m = phone_pattern.search(t)
    if m:
        print(f"Area code: {m.group(1)} from '{t}'")
```
