# Module 008: String Formatting (f-strings, format, %)

> **Phase:** 1. Fundamentals  **Estimated time:** 1.5 hours

Learn to format strings beautifully using f-strings, `.format()`, and legacy `%` formatting.

## f-strings (Python 3.6+)

Prefix the string with `f` and use `{}` for expressions.

```python
name = "Alice"
age = 30
print(f"My name is {name} and I am {age} years old.")

a = 5
b = 3
print(f"{a} + {b} = {a + b}")

print(f"Temperature: {25 * 9/5 + 32}°F")
```

### Format Specifiers

Use `{value:specifier}` to control formatting.

```python
price = 49.9567
print(f"Price: ${price:.2f}")    # 49.96
print(f"Price: ${price:.0f}")    # 50

count = 42
print(f"Count: {count:d}")       # 42

percent = 0.8567
print(f"Score: {percent:.1%}")   # 85.7%
```

### Alignment and Padding

```python
value = 42
print(f"|{value:<10}|")    # left
print(f"|{value:>10}|")    # right
print(f"|{value:^10}|")    # center
print(f"|{value:*^10}|")   # center with * fill
```

## The `.format()` Method

Older but still widely used in existing codebases.

```python
name = "Alice"
age = 30
print("My name is {} and I am {} years old.".format(name, age))

pi = 3.14159
print("Pi is approximately {:.2f}".format(pi))

print("{0} is {1} years old. {0} likes Python.".format(name, age))
```

## Old-Style `%` Formatting

C-style formatting still found in legacy code.

```python
name = "Alice"
age = 30
print("My name is %s and I am %d years old." % (name, age))

price = 49.9567
print("Price: $%.2f" % price)
print("Percent: %.1f%%" % 85.67)
```

## Invoice Formatter Program

```python
item = "Widget A"
quantity = 5
unit_price = 12.345
tax_rate = 0.08

subtotal = quantity * unit_price
tax = subtotal * tax_rate
total = subtotal + tax

print("=== INVOICE ===")
print(f"{'Item':<15} {'Qty':>5} {'Price':>8} {'Total':>8}")
print("-" * 36)
print(f"{item:<15} {quantity:>5} ${unit_price:>6.2f} ${subtotal:>6.2f}")
print()
print(f"{'Subtotal:':>20} ${subtotal:>7.2f}")
print(f"{'Tax (8%):':>20} ${tax:>7.2f}")
print(f"{'Total:':>20} ${total:>7.2f}")
```

## Key Takeaways

- f-strings: `f"{expression}"` — modern, readable, recommended.
- `.format()`: `"{}".format(value)` — legacy but common.
- `%`-formatting: `"%s %d" % (s, n)` — old C-style.
- Specifiers: `:.2f` (2 decimals), `:d` (integer), `:.1%` (percent).
- Alignment: `<` left, `>` right, `^` center.
- Padding: `{value:*>10}` fills with `*`.
- Use f-strings for all new code.
