# Module 068: Working with Dates and Times

```python
from datetime import date, datetime, time, timedelta

# date, time, datetime objects
d = date(2025, 12, 25)
t = time(14, 30, 0)
dt = datetime(2025, 12, 25, 14, 30, 0)

print(f"Date:     {d}")
print(f"Time:     {t}")
print(f"Datetime: {dt}")
```

```python
# timedelta arithmetic
today = date.today()
delta = timedelta(days=7, hours=3)
future = today + delta
past = today - delta
print(f"Today:  {today}")
print(f"Future: {future}")
print(f"Past:   {past}")
```

```python
# strftime: datetime to string
dt = datetime(2025, 12, 25, 10, 30, 0)
formats = [
    "%Y-%m-%d",
    "%m/%d/%Y %I:%M %p",
    "%A, %B %d, %Y",
    "%H:%M:%S",
]
for fmt in formats:
    print(f"'{fmt}' -> {dt.strftime(fmt)}")
```

```python
# strptime: string to datetime
date_strings = [
    "2025-06-15",
    "12/25/2025 10:30 AM",
    "July 4, 2026 at 09:15",
]
formats = [
    "%Y-%m-%d",
    "%m/%d/%Y %I:%M %p",
    "%B %d, %Y at %H:%M",
]
for s, f in zip(date_strings, formats):
    parsed = datetime.strptime(s, f)
    print(f"'{s}' -> {parsed}")
```

```python
# Timezone handling with zoneinfo
from zoneinfo import ZoneInfo

utc = datetime(2025, 6, 15, 12, 0, 0, tzinfo=ZoneInfo("UTC"))
eastern = utc.astimezone(ZoneInfo("US/Eastern"))
pacific = utc.astimezone(ZoneInfo("US/Pacific"))
tokyo = utc.astimezone(ZoneInfo("Asia/Tokyo"))

print(f"UTC:         {utc}")
print(f"US/Eastern:  {eastern}")
print(f"US/Pacific:  {pacific}")
print(f"Asia/Tokyo:  {tokyo}")
```

```python
# Age calculator example
def age(birth_date_str: str) -> int:
    birth = datetime.strptime(birth_date_str, "%Y-%m-%d").date()
    today = date.today()
    return today.year - birth.year - (
        (today.month, today.day) < (birth.month, birth.day)
    )


print(f"Born 1990-05-15: age {age('1990-05-15')}")
print(f"Born 2000-01-01: age {age('2000-01-01')}")
```
