## 1. Nested Dictionaries

```python
company = {
    "Engineering": {"alice": "developer", "bob": "manager"},
    "Marketing": {"charlie": "analyst"}
}
print(company["Engineering"]["alice"])  # developer
```

## 2. Dict Methods

```python
d = {"a": 1, "b": 2, "c": 3}
print("keys:", d.keys())
print("values:", d.values())
print("items:", d.items())
```

```python
# update
d1 = {"a": 1, "b": 2}
d2 = {"b": 3, "c": 4}
d1.update(d2)
print(d1)  # {'a': 1, 'b': 3, 'c': 4}
```

```python
# pop
val = d1.pop("a")
print("Popped:", val, "Remaining:", d1)

# setdefault
d1.setdefault("z", 0)
print(d1)
```

```python
# fromkeys
keys = ["name", "age", "city"]
defaults = dict.fromkeys(keys, None)
print(defaults)
```

## 3. Dict Comprehensions

```python
squares = {x: x**2 for x in range(6)}
print(squares)  # {0: 0, 1: 1, 2: 4, 3: 9, 4: 16, 5: 25}
```

## 4. Merging Dicts with | (Python 3.9+)

```python
a = {"x": 1, "y": 2}
b = {"y": 3, "z": 4}
merged = a | b
print(merged)  # {'x': 1, 'y': 3, 'z': 4}
```

## 5. defaultdict

```python
from collections import defaultdict

# Group words by first letter
words = ["apple", "banana", "avocado", "cherry"]
groups = defaultdict(list)
for w in words:
    groups[w[0]].append(w)

print(dict(groups))
```

## Summary
- Nest dicts for structured data
- Use `.update()`, `.pop()`, `.setdefault()`, `.fromkeys()`
- Dict comprehensions: `{k: v for ...}`
- Merge with `|` operator
- `defaultdict` provides automatic default values
