# Module 051: Composition vs Inheritance

Learn the difference between "has-a" (composition) and "is-a" (inheritance) relationships.

````python
# Inheritance: "is-a" relationship
class Vehicle:
    def __init__(self, make: str, model: str) -> None:
        self.make = make
        self.model = model

    def start(self) -> str:
        return f"{self.make} {self.model} engine started"


class ElectricCar(Vehicle):
    """ElectricCar is-a Vehicle."""
    def start(self) -> str:
        return f"{self.make} {self.model} silently powered up"


tesla = ElectricCar("Tesla", "Model 3")
print(tesla.start())
````

````python
# Composition: "has-a" relationship
class Engine:
    def __init__(self, horsepower: int) -> None:
        self.horsepower = horsepower

    def start(self) -> str:
        return f"Engine ({self.horsepower} HP) started"


class Car:
    """Car has-an Engine."""
    def __init__(self, make: str, model: str, engine: Engine) -> None:
        self.make = make
        self.model = model
        self.engine = engine  # composition

    def start(self) -> str:
        return f"{self.make} {self.model}: {self.engine.start()}"


v8 = Engine(450)
mustang = Car("Ford", "Mustang", v8)
print(mustang.start())
````

````python
# Favor composition over inheritance
from typing import List

class Logger:
    def log(self, message: str) -> None:
        print(f"[LOG] {message}")


class EmailSender:
    def send(self, to: str, subject: str) -> None:
        print(f"Email sent to {to}: {subject}")


class UserService:
    """Uses composition to combine logging and email capabilities."""
    def __init__(self) -> None:
        self.logger = Logger()
        self.emailer = EmailSender()

    def register(self, email: str) -> None:
        self.logger.log(f"Registering user {email}")
        self.emailer.send(email, "Welcome!")


svc = UserService()
svc.register("alice@example.com")
````

## UML Diagram

```
[Vehicle]           [Car]  has-a  [Engine]
    ^                        |
    |                        |
[ElectricCar]    [UserService] has-a [Logger, EmailSender]

Inheritance (is-a):   ElectricCar --|> Vehicle
Composition (has-a):  Car o-- Engine
                      UserService o-- Logger
                      UserService o-- EmailSender
```
