Instance attributes store data; methods define behavior.

## Instance Attributes and Methods

Attributes are created with `self.attr` in `__init__`.

```python
class Student:
    """Represent a student."""

    def __init__(self, name, student_id):
        self.name = name
        self.student_id = student_id
        self.courses = []

    def enroll(self, course):
        self.courses.append(course)

    def display_info(self):
        info = f"{self.name} (ID: {self.student_id})\n"
        info += f"Enrolled: {', '.join(self.courses) if self.courses else 'None'}"
        return info


alice = Student("Alice", "S1001")
alice.enroll("Python 101")
alice.enroll("Data Structures")
print(alice.display_info())
```

## Adding / Changing Attributes Dynamically

```python
alice = Student("Alice", "S1001")
print(f"Before: {alice.name}")

alice.name = "Alice B."  # change existing
alice.gpa = 3.8           # add new dynamically

print(f"After: {alice.name}")
print(f"GPA: {alice.gpa}")
```

## Method vs Function

```python
def greet(name):          # standalone function
    return f"Hello {name}"


class Person:
    def greet(self):      # method
        return f"Hello {self.name}"


print(greet("World"))          # function call

p = Person()
p.name = "World"
print(p.greet())                # method call

# Methods are bound to the instance:
print(Person.greet(p))          # equivalent explicit call
```

## BankAccount Example

```python
class BankAccount:
    """A simple bank account."""

    def __init__(self, owner, balance=0):
        self.owner = owner
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount

    def withdraw(self, amount):
        if amount > self.balance:
            print("Insufficient funds!")
        else:
            self.balance -= amount

    def info(self):
        return f"{self.owner}: ${self.balance}"


acc = BankAccount("Alice", 100)
acc.deposit(50)
acc.withdraw(30)
print(acc.info())
acc.account_type = "checking"
print(f"Type: {acc.account_type}")
```

## Key Takeaways

- Attributes store data; methods define behavior
- Use `self.attr` in `__init__` to initialize attributes
- Methods are called on an instance: `instance.method()`
- Python allows dynamic attribute addition
- Methods are functions that belong to an instance
