## pyproject.toml Example

````python
content: str = '''
[build-system]
requires = ["setuptools>=61.0"]
build-backend = "setuptools.build_meta"

[project]
name = "sample-toolkit"
version = "0.1.0"
description = "A sample Python package for demonstration"
readme = "README.md"
license = {text = "MIT"}
keywords = ["sample", "toolkit", "demo"]
classifiers = [
    "Development Status :: 3 - Alpha",
    "Intended Audience :: Developers",
    "License :: OSI Approved :: MIT License",
    "Programming Language :: Python :: 3",
    "Programming Language :: Python :: 3.10",
    "Programming Language :: Python :: 3.11",
    "Programming Language :: Python :: 3.12",
]
dependencies = [
    "requests>=2.28",
]

[project.urls]
Homepage = "https://github.com/username/sample-toolkit"
Documentation = "https://sample-toolkit.readthedocs.io"

[project.scripts]
sample-tool = "sample_toolkit.cli:main"
'''
print(content)
````

## __version__ Convention

````python
# In sample_toolkit/__init__.py
__version__: str = "0.1.0"

# Accessing version
from importlib.metadata import version
try:
    ver: str = version("sample-toolkit")
    print(f"Installed version: {ver}")
except Exception:
    print("Package not installed (expected in notebook demo)")

print(f"Direct __version__: sample_toolkit.__version__")
````

## Building and Publishing (CLI Commands)

````python
build_commands: str = '''
# Install build tools
pip install build twine

# Build source distribution and wheel
python -m build

# Check distribution
twine check dist/*

# Upload to Test PyPI
twine upload --repository-url https://test.pypi.org/legacy/ dist/*

# Upload to PyPI
twine upload dist/*

# Install from Test PyPI
pip install --index-url https://test.pypi.org/simple/ sample-toolkit
'''
print(build_commands)
````

## Semantic Versioning (SemVer)

```
MAJOR.MINOR.PATCH
   |      |     |
   |      |     +-- Patch: bug fixes (backward-compatible)
   |      +-------- Minor: new features (backward-compatible)
   +--------------- Major: breaking changes

Example: 2.1.3 → Major=2, Minor=1, Patch=3

Pre-release: 1.0.0-alpha, 1.0.0-beta.2
Build metadata: 1.0.0+build.20230101
```
