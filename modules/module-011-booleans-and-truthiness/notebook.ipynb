# Module 011: Booleans and Truthiness

Learn how Python represents true/false values and what makes other values "truthy" or "falsy".

## 1. Boolean Values

The `bool` type has exactly two values: `True` and `False`.

```python
is_sunny = True
is_raining = False
print(f"is_sunny: {is_sunny}, type: {type(is_sunny)}")
print(f"is_raining: {is_raining}, type: {type(is_raining)}")
```

```python
# bool is a subclass of int
print(f"True + True = {True + True}")
print(f"True * 5 = {True * 5}")
print(f"False * 10 = {False * 10}")
```

## 2. The `bool()` Function

Use `bool()` to convert any value to a Boolean.

```python
print(f"bool(1): {bool(1)}")
print(f"bool(0): {bool(0)}")
print(f"bool(3.14): {bool(3.14)}")
print(f"bool(0.0): {bool(0.0)}")
print(f"bool('hello'): {bool('hello')}")
print(f"bool(''): {bool('')}")
print(f"bool(None): {bool(None)}")
print(f"bool([]): {bool([])}")
print(f"bool([1, 2]): {bool([1, 2])}")
```

## 3. Truthiness Check

Values that are falsy: `0`, `0.0`, `""`, `None`, `[]`, `False`. Everything else is truthy.

```python
test_values = [0, 1, 0.0, -5, "", "hello", None, "False", [], [0], False, True]
for val in test_values:
    print(f"bool({val!r:8}) -> {bool(val)}")
```

## 4. Boolean Operators

### `not` — Logical NOT

```python
print(f"not True: {not True}")
print(f"not False: {not False}")
print(f"not 0: {not 0}")
print(f"not 42: {not 42}")
print(f"not '': {not ''}")
print(f"not 'hello': {not 'hello'}")
```

### `and` — Logical AND

```python
print(f"True and True: {True and True}")
print(f"True and False: {True and False}")
print(f"False and True: {False and True}")
print(f"False and False: {False and False}")
print()
print(f"0 and 42: {0 and 42}")
print(f"3 and 7: {3 and 7}")
print(f"'' and 'hi': {'' and 'hi'}")
print(f"'hi' and 'there': {'hi' and 'there'}")
```

### `or` — Logical OR

```python
print(f"True or True: {True or True}")
print(f"True or False: {True or False}")
print(f"False or True: {False or True}")
print(f"False or False: {False or False}")
print()
print(f"0 or 42: {0 or 42}")
print(f"3 or 7: {3 or 7}")
print(f"'' or 'hi': {'' or 'hi'}")
print(f"0 or None: {0 or None}")
```

## 5. Short-Circuit Evaluation

Python stops evaluating as soon as the result is known.

```python
def noisy_check(value):
    print(f"  checking {value}...")
    return value > 0

print("and with first falsy (short-circuits):")
result = noisy_check(5) and noisy_check(-1) and noisy_check(10)
print(f"Result: {result}")
```

```python
def get_name(index):
    names = ["", "", "Alice", "Bob"]
    print(f"  get_name({index}) called")
    return names[index]

print("or with first truthy (short-circuits):")
name = get_name(0) or get_name(1) or get_name(2) or get_name(3)
print(f"Name: {name}")
```

## 6. Practical Examples

```python
# Using truthiness for input validation
username = ""  # simulate empty input
if not username:
    print("Username cannot be empty!")

score = 0
if not score:
    print("Score is zero — needs attention.")

# Combinining conditions
age = 25
has_ticket = True
if age >= 18 and has_ticket:
    print("Welcome to the show!")
```

```python
# Default value pattern with or
user_input = ""  # empty input
name = user_input or "Guest"
print(f"Hello, {name}!")

# Multiple conditions
x = 7
print(f"x > 0 and x < 10: {x > 0 and x < 10}")
print(f"x < 0 or x > 5: {x < 0 or x > 5}")
```
