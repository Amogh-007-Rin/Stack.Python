## Basic Flask App

````python
from flask import Flask, request, redirect, url_for, render_template, jsonify

app: Flask = Flask(__name__)
````

## Routes and Dynamic URLs

````python
@app.route('/')
def home() -> str:
    return '<h1>Welcome to Flask!</h1>'

@app.route('/user/<name>')
def greet(name: str) -> str:
    return f'<h1>Hello, {name}!</h1>'

@app.route('/search')
def search() -> str:
    q: str = request.args.get('q', '')
    return f'<p>Search results for: {q}</p>' if q else '<p>No query provided.</p>'

# Test with app.test_client
with app.test_client() as client:
    print(client.get('/').data.decode())
    print(client.get('/user/Alice').data.decode())
    print(client.get('/search?q=flask').data.decode())
````

## GET and POST Methods

````python
@app.route('/contact', methods=['GET', 'POST'])
def contact() -> str:
    if request.method == 'POST':
        name: str = request.form.get('name', 'Anonymous')
        return redirect(url_for('thank_you', name=name))
    return '''
        <form method="post">
            <input type="text" name="name" placeholder="Your name">
            <button type="submit">Send</button>
        </form>
    '''

@app.route('/thank-you')
def thank_you() -> str:
    name: str = request.args.get('name', 'Guest')
    return f'<h2>Thanks, {name}!</h2>'

with app.test_client() as client:
    print(client.get('/contact').data.decode()[:100])
    resp = client.post('/contact', data={'name': 'Alice'}, follow_redirects=True)
    print('After POST:', resp.data.decode())
````

## JSON Response

````python
@app.route('/api/data')
def api_data() -> dict:
    return jsonify({'app': 'Flask Demo', 'version': '1.0', 'status': 'ok'})

with app.test_client() as client:
    print(client.get('/api/data').data.decode())
````
