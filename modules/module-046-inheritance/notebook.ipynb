## Single Inheritance

A child class inherits everything from its parent.

```python
class Animal:
    """Base class."""

    def __init__(self, name):
        self.name = name

    def speak(self):
        return f"{self.name} makes a sound."


class Dog(Animal):
    """Dog inherits from Animal."""

    def speak(self):
        return f"{self.name} says Woof!"


class Cat(Animal):
    """Cat inherits from Animal."""

    def speak(self):
        return f"{self.name} says Meow!"


animals = [Dog("Rex"), Cat("Luna"), Animal("Generic")]
for a in animals:
    print(a.speak())
```

## Using `super()`

Call parent methods from a child class.

```python
class Vehicle:
    """Base vehicle class."""

    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year

    def description(self):
        return f"{self.year} {self.make} {self.model}"


class ElectricCar(Vehicle):
    """Electric car extends Vehicle."""

    def __init__(self, make, model, year, battery_kwh):
        super().__init__(make, model, year)
        self.battery_kwh = battery_kwh

    def description(self):
        base = super().description()
        return f"{base} ({self.battery_kwh} kWh)"


tesla = ElectricCar("Tesla", "Model 3", 2024, 75)
print(tesla.description())
```

## `isinstance()` and `issubclass()`

```python
animal = Animal("Generic")
dog = Dog("Rex")

print(f"dog is Dog: {isinstance(dog, Dog)}")
print(f"dog is Animal: {isinstance(dog, Animal)}")
print(f"animal is Dog: {isinstance(animal, Dog)}")
print(f"Dog subclass Animal: {issubclass(Dog, Animal)}")
```

## Method Resolution Order (MRO)

```python
class A:
    def who(self):
        return "A"


class B(A):
    def who(self):
        return "B"


class C(A):
    def who(self):
        return "C"


class D(B, C):
    pass


d = D()
print(f"D.who(): {d.who()}")
print(f"MRO: {[c.__name__ for c in D.__mro__]}")
```

## Inheriting from Built-in Types

```python
class UpperCaseDict(dict):
    """Dict that uppercases all keys."""

    def __setitem__(self, key, value):
        super().__setitem__(key.upper(), value)


d = UpperCaseDict()
d["name"] = "Alice"
d["city"] = "Boston"
print(d)  # {'NAME': 'Alice', 'CITY': 'Boston'}
```

## Key Takeaways

- Child classes inherit parent attributes and methods
- Override methods to specialize behavior
- `super()` calls the parent's implementation
- `isinstance()` and `issubclass()` check type relationships
- MRO determines method lookup order
