# Module 033: Scope and Namespaces (the LEGB Rule)

> **Phase:** 4. Functions  **Estimated time:** 1.5 hours

Understand how Python looks up variable names.

## Local Scope

Variables inside a function are local:

```python
def my_func():
    x = 5
    print(f"Inside: x = {x}")

my_func()
try:
    print(x)
except NameError as e:
    print(f"Outside: {e}")
```

## Global Scope

Variables at the module top level are global:

```python
y = 10

def show_y():
    print(f"Inside function: y = {y}")

show_y()
print(f"Outside function: y = {y}")
```

## The `global` Keyword

Use `global` to modify a global variable:

```python
counter = 0

def increment():
    global counter
    counter += 1

increment()
increment()
increment()
print(f"Counter after 3 increments: {counter}")
```

## Enclosing Scope and `nonlocal`

Nested functions have an enclosing scope:

```python
def outer():
    x = "outer"
    def inner():
        nonlocal x
        x = "inner modified"
    inner()
    print(f"After inner: x = {x}")

outer()
```

## Shadowing

A local variable shadows outer variables with the same name:

```python
x = 100

def shadow():
    x = 200  # local — shadows the global
    print(f"Inside: x = {x}")

shadow()
print(f"Outside: x = {x}")
```

## Built-in Scope

Names like `print`, `len`, `range` are in the built-in scope:

```python
print("print is a built-in function")
print(f"len('hello') = {len('hello')}")
print(f"list(range(5)) = {list(range(5))}")
```

## LEGB Lookup in Action

Python checks Local → Enclosing → Global → Built-in:

```python
z = "global z"

def outer():
    z = "enclosing z"
    def inner():
        z = "local z"
        print(f"Inner sees: {z}")
    inner()

outer()
print(f"Global: {z}")
```

## Key Takeaways

- LEGB: Local → Enclosing → Global → Built-in.
- `global` modifies module-level variables inside a function.
- `nonlocal` modifies enclosing function variables.
- Shadowing happens when a local name hides an outer name.
- Minimize global variables for cleaner code.
