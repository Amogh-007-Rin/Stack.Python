## String Indexing

Zero-based indexing. Negative indices count from the end.

```python
word = "Python"
print("First char:", word[0])
print("Third char:", word[2])
print("Last char:", word[-1])
print("Second from end:", word[-2])
```

## String Slicing

`string[start:stop:step]` — stop is exclusive.

```python
text = "Hello, World!"

print("text[0:5]:", text[0:5])     # Hello
print("text[7:12]:", text[7:12])   # World
print("text[:5]:", text[:5])       # Hello
print("text[7:]:", text[7:])       # World!
print("text[::2]:", text[::2])     # Hlo ol!
print("text[::-1]:", text[::-1])   # !dlroW ,olleH
```

## len() Function

```python
print("len('Python'):", len("Python"))
print("len(''):", len(""))
print("len('Hello\n'):", len("Hello\n"))
```

## Common String Methods

### Case Conversion

```python
text = "  Hello, Python World!  "
print("upper:", text.upper())
print("lower:", text.lower())
print("strip:", text.strip())
print("replace:", text.replace("World", "Universe"))
```

### Searching

```python
print("find 'Python':", text.find("Python"))
print("find 'Java':", text.find("Java"))
print("count 'o':", text.count("o"))
print("'Python' in text:", "Python" in text)
```

### Split and Join

```python
csv = "apple,banana,cherry"
fruits = csv.split(",")
print("split:", fruits)

joined = " - ".join(fruits)
print("join:", joined)
```

## String Immutability

Strings cannot be changed in place. Methods return new strings.

```python
name = "alice"
name.upper()
print("Original unchanged:", name)  # alice

name = name.upper()
print("Reassigned:", name)  # ALICE
```

## Escape Sequences

```python
print("Hello\nWorld")           # newline
print("Tab\tseparated")         # tab
print("C:\\Users\\name")      # backslashes
print('It\'s nice')              # single quote
print("He said \"Hi\"")        # double quote
```

## Multi-line Strings

```python
poem = """Roses are red,
Violets are blue,
Python is awesome,
And so are you."""
print(poem)
```

## Data Cleaner Program

Process a messy data string using multiple string methods.

```python
raw = "  John;DOE;35;New York  "
clean = raw.strip()
parts = clean.split(";")

first = parts[0].capitalize()
last = parts[1].capitalize()
age = parts[2]
city = parts[3]

print("Name:", first, last)
print("Age:", age)
print("City:", city)

initials = first[0] + "." + last[0] + "."
print("Initials:", initials)

sentence = "The quick brown fox jumps over the lazy dog"
words = sentence.split()
print("Word count:", len(words))
print("'The' appears", sentence.lower().count("the"), "times")
```

## Key Takeaways

- Index: `s[0]` first, `s[-1]` last.
- Slice: `s[start:stop:step]`, stop exclusive.
- `s[::-1]` reverses a string.
- `len(s)` returns character count.
- Methods: `.upper()`, `.lower()`, `.strip()`, `.split()`, `.join()`, `.replace()`, `.find()`, `.count()`.
- Strings are immutable.
- Escape sequences: `\n`, `\t`, `\\`, `\'`, `\"`.
- Triple quotes for multi-line strings.
