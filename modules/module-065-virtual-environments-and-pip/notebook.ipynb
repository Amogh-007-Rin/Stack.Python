# Module 065: Virtual Environments and pip

This module covers virtual environments and pip. Most commands run in the terminal, not in the notebook.

## Key Commands

- **Creating a venv:** `python -m venv myenv`
- **Activate (Linux/macOS):** `source myenv/bin/activate`
- **Activate (Windows):** `myenv\\Scripts\\activate`
- **Install package:** `pip install requests`
- **Freeze requirements:** `pip freeze > requirements.txt`
- **Install from file:** `pip install -r requirements.txt`
- **List packages:** `pip list`
- **Uninstall:** `pip uninstall requests`
- **Deactivate:** `deactivate`

```python
# Check if we're in a virtual environment
import sys
print(f"Python executable: {sys.executable}")
print(f"In venv: {hasattr(sys, 'real_prefix') or (hasattr(sys, 'base_prefix') and sys.base_prefix != sys.prefix)}")
```

```python
# Simulate requirements.txt content
reqs = """flask==2.3.0
requests==2.28.0
numpy==1.24.0
"""

with open("requirements.txt", "w") as f:
    f.write(reqs)

print("requirements.txt created:")
with open("requirements.txt", "r") as f:
    print(f.read())
```

```python
# Check installed packages via pip
import subprocess
result = subprocess.run([sys.executable, "-m", "pip", "list", "--format=columns"],
                       capture_output=True, text=True)
print(result.stdout[:500])
```
