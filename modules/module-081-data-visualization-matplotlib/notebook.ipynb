## Line Plots with plt.plot

```python
import matplotlib.pyplot as plt

x: list[int] = [1, 2, 3, 4, 5]
y: list[int] = [1, 4, 9, 16, 25]
plt.plot(x, y, marker='o')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Line Plot Example')
plt.show()
```

## Bar Charts

```python
categories: list[str] = ['Apples', 'Bananas', 'Cherries', 'Dates']
values: list[int] = [25, 40, 30, 15]
plt.bar(categories, values, color=['red', 'blue', 'green', 'orange'])
plt.title('Fruit Sales')
plt.ylabel('Quantity')
plt.show()

# Horizontal bar chart
plt.barh(categories, values, color='skyblue')
plt.title('Fruit Sales (Horizontal)')
plt.xlabel('Quantity')
plt.show()
```

## Scatter Plots and Histograms

```python
import numpy as np

rng: np.random.Generator = np.random.default_rng(42)
x: np.ndarray = rng.random(50) * 10
y: np.ndarray = 2 * x + rng.normal(0, 1, 50) * 2
plt.scatter(x, y, alpha=0.7, c='purple', marker='^')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot with Noise')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
data: np.ndarray = rng.randn(1000)
plt.hist(data, bins=30, edgecolor='black', alpha=0.7)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Normal Distribution')
plt.show()
```

## Pie Charts

```python
labels: list[str] = ['Python', 'JavaScript', 'Java', 'C++', 'Other']
sizes: list[int] = [35, 25, 20, 10, 10]
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, explode=(0.1, 0, 0, 0, 0))
plt.title('Programming Language Popularity')
plt.show()
```

## Subplots

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

x: np.ndarray = np.linspace(0, 10, 100)
axes[0, 0].plot(x, np.sin(x))
axes[0, 0].set_title('Sine')

axes[0, 1].bar(['A', 'B', 'C'], [3, 7, 2], color='green')
axes[0, 1].set_title('Bar')

axes[1, 0].scatter(rng.random(30), rng.random(30), c='red')
axes[1, 0].set_title('Scatter')

axes[1, 1].hist(rng.randn(500), bins=20, edgecolor='black')
axes[1, 1].set_title('Histogram')

fig.suptitle('Multiple Subplots', fontsize=14)
plt.tight_layout()
plt.show()
```
