## Implicit Conversion

Python automatically converts types when there is no risk of data loss.

```python
result = 3 + 2.0
print(result, type(result))

total = 10 + True
print(total, type(total))
```

Implicit conversion has limits. Python won't guess about strings:

```python
# This causes an error — uncomment to see:
# print("42" + 3)
```

## Explicit Conversion (Casting)

Use `int()`, `float()`, `str()`, and `bool()` to convert explicitly.

```python
# int()
print(int(3.99))    # 3 — truncates
print(int("42"))    # 42
print(int(True))    # 1
```

```python
# float()
print(float(3))       # 3.0
print(float("3.14"))  # 3.14
print(float(True))    # 1.0
```

```python
# str()
print(str(42))     # "42"
print(str(3.14))   # "3.14"
print(str(False))  # "False"
```

```python
# bool() — truthiness
print(bool(1))       # True
print(bool(0))       # False
print(bool(""))     # False
print(bool("Hi"))   # True
print(bool(None))    # False
```

## Conversion Errors

Not all strings can be converted.

```python
# Uncomment to see the errors:
# int("hello")   # ValueError
# int("3.14")    # ValueError — convert to float first
# float("3,14")  # ValueError — use dot, not comma
```

### Safe conversion: convert to float first, then to int

```python
value = "3.14"
result = int(float(value))
print(result, type(result))  # 3 <class 'int'>
```

## Hands-On: Simulated Data Conversion

Imagine we read data from a file (all strings) and need proper types.

```python
age_string = "25"
height_string = "1.75"
has_license = "True"

age = int(age_string)
height = float(height_string)
license_bool = bool(has_license)

print("Age:", age, type(age))
print("Height:", height, type(height))
print("Has license:", license_bool, type(license_bool))
print("Next year:", age + 1)
print("You are " + str(age) + " years old.")
```

## Key Takeaways

- Implicit conversion: int → float, int → bool (automatic).
- Explicit conversion: `int()`, `float()`, `str()`, `bool()`.
- `int()` truncates toward zero (does not round).
- `bool("")` and `bool(0)` are `False`; almost everything else is `True`.
- Convert `"3.14"` to `float` first, then to `int` to avoid errors.
- Always cast explicitly — never rely on guesses.
