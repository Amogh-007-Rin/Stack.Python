## Creating Series and DataFrames

```python
import pandas as pd

# Series
ages = pd.Series([25, 30, 35], name='Age')
print('Series:')
print(ages)

# DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'City': ['NYC', 'LA', 'Chicago']
}
df = pd.DataFrame(data)
print('\nDataFrame:')
print(df)
```

## Reading Data

```python
import pandas as pd
from io import StringIO

csv_data = """Name,Age,City
Alice,25,NYC
Bob,30,LA
Charlie,35,Chicago
"""

df = pd.read_csv(StringIO(csv_data))
print('From CSV:')
print(df)
```

## Exploring Data

```python
import pandas as pd

df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'Age': [25, 30, 35, 28, 22],
    'Salary': [70000, 80000, 90000, 75000, 65000]
})

print('Head:')
print(df.head())
print('\nInfo:')
print(df.info())
print('\nDescribe:')
print(df.describe())
print('\nShape:', df.shape)
print('Columns:', df.columns.tolist())
```

## Selecting and Filtering

```python
import pandas as pd

df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana'],
    'Age': [25, 30, 35, 28],
    'City': ['NYC', 'LA', 'Chicago', 'NYC'],
    'Salary': [70000, 80000, 90000, 75000]
})

# Select columns
print('Names and Salaries:')
print(df[['Name', 'Salary']])

# Filter rows
high_earners = df[df['Salary'] > 75000]
print('\nHigh earners:')
print(high_earners)

# Multiple conditions
result = df[(df['Age'] > 25) & (df['City'] == 'NYC')]
print('\nAge > 25 and NYC:')
print(result)
```

## Basic Calculations

```python
import pandas as pd

df = pd.DataFrame({
    'Sales': [100, 200, 150, 300, 250],
    'Costs': [80, 150, 120, 220, 200]
})

df['Profit'] = df['Sales'] - df['Costs']
print('With profit column:')
print(df)

print('\nTotal Sales:', df['Sales'].sum())
print('Mean Profit:', df['Profit'].mean())
print('Max Costs:', df['Costs'].max())
print('Min Sales:', df['Sales'].min())
```
