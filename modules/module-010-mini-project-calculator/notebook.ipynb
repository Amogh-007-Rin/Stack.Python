## Project Overview

The calculator should:
1. Prompt for two numbers
2. Prompt for an operator (`+`, `-`, `*`, `/`, `//`, `%`, `**`)
3. Compute and display the result
4. Handle division by zero
5. Loop until the user types "done"
6. Format output cleanly

## Step 1: Get User Input

Let's start with the basic loop and input handling.

```python
# Simulated version (since input() may not work in all notebook environments)
print("=== Python Command-Line Calculator ===")
print("Enter 'done' to exit.")
print()

# We'll simulate one calculation
num1 = 10.0
operator = "/"
num2 = 3.0

print(f"First number: {num1}")
print(f"Operator: {operator}")
print(f"Second number: {num2}")
```

## Step 2: Perform the Calculation

Use conditional logic to apply the correct operation.

```python
if operator == "+":
    result = num1 + num2
elif operator == "-":
    result = num1 - num2
elif operator == "*":
    result = num1 * num2
elif operator == "/":
    if num2 == 0:
        result = "Error: Division by zero is not allowed."
    else:
        result = num1 / num2
elif operator == "//":
    if num2 == 0:
        result = "Error: Division by zero is not allowed."
    else:
        result = num1 // num2
elif operator == "%":
    if num2 == 0:
        result = "Error: Division by zero is not allowed."
    else:
        result = num1 % num2
elif operator == "**":
    result = num1 ** num2
else:
    result = "Error: Unknown operator."

print(f"{num1} {operator} {num2} = {result}")
```

## Step 3: The Complete Calculator (Simulated Loops)

Here's the full program running through several examples:

```python
def calculate(num1, operator, num2):
    if operator == "+":
        return num1 + num2
    elif operator == "-":
        return num1 - num2
    elif operator == "*":
        return num1 * num2
    elif operator == "/":
        if num2 == 0:
            return "Error: Division by zero is not allowed."
        return num1 / num2
    elif operator == "//":
        if num2 == 0:
            return "Error: Division by zero is not allowed."
        return num1 // num2
    elif operator == "%":
        if num2 == 0:
            return "Error: Division by zero is not allowed."
        return num1 % num2
    elif operator == "**":
        return num1 ** num2
    else:
        return "Error: Unknown operator."

print("=== Python Command-Line Calculator ===")
print()

test_cases = [
    (10.0, "/", 3.0),
    (10.0, "/", 0.0),
    (2.0, "**", 10.0),
    (15.0, "//", 4.0),
    (17.0, "%", 5.0),
    (8.0, "*", 7.0),
]

for n1, op, n2 in test_cases:
    res = calculate(n1, op, n2)
    print(f"{n1} {op} {n2} = {res}")

print()
print("Goodbye!")
```
