## Naming Conventions

Python uses conventions rather than strict access control.

```python
class BankAccount:
    """Bank account with encapsulated balance."""

    def __init__(self, owner, balance):
        self.owner = owner
        self._branch_code = "001"     # protected (convention)
        self.__balance = balance      # name mangled

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount

    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount
            return amount
        raise ValueError("Invalid withdrawal")

    def get_balance(self):
        return self.__balance


acc = BankAccount("Alice", 1000)
print(f"Owner: {acc.owner}")
print(f"Balance: {acc.get_balance()}")

# Access conventions:
print(f"Branch (protected): {acc._branch_code}")  # works but discouraged
# print(acc.__balance)  # AttributeError!
print(f"Mangled: {acc._BankAccount__balance}")    # accessible but clearly internal
```

## Getter / Setter Pattern

Explicit methods for controlled access.

```python
class Temperature:
    """Temperature with validation."""

    def __init__(self, celsius=0):
        self._celsius = celsius

    def get_celsius(self):
        return self._celsius

    def set_celsius(self, value):
        if value < -273.15:
            raise ValueError("Below absolute zero!")
        self._celsius = value

    def get_fahrenheit(self):
        return self._celsius * 9 / 5 + 32


t = Temperature(25)
print(f"Celsius: {t.get_celsius()}")
print(f"Fahrenheit: {t.get_fahrenheit():.1f}")
t.set_celsius(30)
print(f"Updated: {t.get_celsius()}C / {t.get_fahrenheit():.1f}F")

try:
    t.set_celsius(-300)
except ValueError as e:
    print(f"Error: {e}")
```

## Why Encapsulation Matters

1. **Validation**: Prevent invalid state
2. **Decoupling**: Change internals without breaking users
3. **Debugging**: Add logging at method boundaries
4. **Contracts**: Methods document what goes in/out

```python
class Person:
    """Person with encapsulated age."""

    def __init__(self, name, age):
        self.name = name
        self.__age = age

    def get_age(self):
        return self.__age

    def set_age(self, age):
        if not (0 < age < 150):
            raise ValueError("Invalid age")
        self.__age = age


p = Person("Alice", 30)
print(f"{p.name} is {p.get_age()} years old")

try:
    p.set_age(200)
except ValueError as e:
    print(f"Validation works: {e}")
```

## Key Takeaways

- `_name` = protected convention; `__name` = name mangling
- Getters/setters enable validation and future flexibility
- Encapsulation makes code more robust and maintainable
- Python trusts developers — conventions, not enforcement
