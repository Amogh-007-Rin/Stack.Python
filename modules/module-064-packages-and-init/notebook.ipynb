# Module 064: Packages and \_\_init\_\_.py

```python
# Creating a package structure
import os
os.makedirs("mypackage", exist_ok=True)

# Create __init__.py
with open("mypackage/__init__.py", "w") as f:
    f.write('''
from .math_ops import add, subtract

__all__ = ["add", "subtract"]
''')

# Create math_ops.py module
with open("mypackage/math_ops.py", "w") as f:
    f.write('''
def add(a: float, b: float) -> float:
    return a + b

def subtract(a: float, b: float) -> float:
    return a - b

def multiply(a: float, b: float) -> float:
    return a * b
''')

print("Package created")
```

```python
# Importing from the package
from mypackage import add, subtract
from mypackage.math_ops import multiply

print(f"add(5, 3) = {add(5, 3)}")
print(f"subtract(5, 3) = {subtract(5, 3)}")
print(f"multiply(5, 3) = {multiply(5, 3)}")
```

```python
# Subpackages
os.makedirs("mypackage/advanced", exist_ok=True)

with open("mypackage/advanced/__init__.py", "w") as f:
    pass

with open("mypackage/advanced/calc.py", "w") as f:
    f.write('''
from ..math_ops import add

def factorial(n: int) -> int:
    if n <= 1:
        return 1
    return n * factorial(n - 1)
''')

from mypackage.advanced.calc import factorial
print(f"factorial(5) = {factorial(5)}")
```
