# Module 057: Custom Exceptions

```python
# Creating custom exceptions
class InsufficientFundsError(Exception):
    """Raised when an account has insufficient funds."""
    def __init__(self, balance: float, amount: float) -> None:
        self.balance = balance
        self.amount = amount
        super().__init__(f"Insufficient funds: balance=${balance:.2f}, needed=${amount:.2f}")


class BankAccount:
    def __init__(self, owner: str, balance: float = 0.0) -> None:
        self.owner = owner
        self.balance = balance

    def withdraw(self, amount: float) -> float:
        if amount > self.balance:
            raise InsufficientFundsError(self.balance, amount)
        self.balance -= amount
        return amount


acct = BankAccount("Alice", 100.0)
try:
    acct.withdraw(200.0)
except InsufficientFundsError as e:
    print(f"Error: {e}")
```

```python
# Exception hierarchy
class PaymentError(Exception):
    """Base exception for payment system."""


class CardDeclinedError(PaymentError):
    """Card was declined by processor."""


class ExpiredCardError(PaymentError):
    """Card has expired."""


class InsufficientFunds(PaymentError):
    """Card has insufficient funds."""


def process_payment(card_number: str, amount: float) -> str:
    if len(card_number) != 16:
        raise CardDeclinedError("Invalid card number")
    if amount > 1000:
        raise InsufficientFunds("Amount exceeds limit")
    return "Payment approved"


try:
    process_payment("1234", 100)
except PaymentError as e:
    print(f"Payment failed: {type(e).__name__}: {e}")
```
