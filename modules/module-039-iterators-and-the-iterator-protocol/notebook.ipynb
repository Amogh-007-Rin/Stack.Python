## iter() and next()

Manual iteration using the protocol:

```python
my_list = [10, 20, 30]
it = iter(my_list)

print(next(it))
print(next(it))
print(next(it))

try:
    print(next(it))
except StopIteration:
    print("Iterator exhausted!")
```

## For Loop Internals

A `for` loop manually calls iter() and next():

```python
# What Python does internally:
def for_loop(iterable):
    """Simulate a for loop using the iterator protocol."""
    _it = iter(iterable)
    while True:
        try:
            item = next(_it)
        except StopIteration:
            break
        print(item)

print("Simulated for loop:")
for_loop([1, 2, 3])
```

## Custom Iterator Class

Implement `__iter__` and `__next__`:

```python
class CountDown:
    """Iterate from n down to 1."""

    def __init__(self, n):
        self.n = n

    def __iter__(self):
        return self

    def __next__(self):
        if self.n <= 0:
            raise StopIteration
        value = self.n
        self.n -= 1
        return value

for i in CountDown(5):
    print(i, end=" ")
```

## Custom Iterator: EvenNumbers

Another example with even numbers:

```python
class EvenNumbers:
    """Iterate through even numbers up to max_n."""

    def __init__(self, max_n):
        self.current = 0
        self.max_n = max_n

    def __iter__(self):
        return self

    def __next__(self):
        self.current += 2
        if self.current > self.max_n:
            raise StopIteration
        return self.current

evens = EvenNumbers(10)
print(list(evens))  # [2, 4, 6, 8, 10]
```

## Iterable vs Iterator

An iterable can produce iterators; an iterator produces values:

```python
nums = [1, 2, 3]  # iterable (has __iter__, no __next__)
it = iter(nums)   # iterator (has both __iter__ and __next__)

print(f"List is iterable: {hasattr(nums, '__iter__')}")
print(f"List has __next__: {hasattr(nums, '__next__')}")
print(f"Iterator has __next__: {hasattr(it, '__next__')}")

print("\nIterable can produce multiple iterators:")
it1 = iter(nums)
it2 = iter(nums)
print(list(it1))  # [1, 2, 3]
print(list(it2))  # [1, 2, 3]
```

## Key Takeaways

- Iterator protocol: `__iter__()` returns self, `__next__()` returns next value or raises `StopIteration`.
- `for` loop = `iter()` + `next()` + `StopIteration`.
- Custom iterators are classes that implement the protocol.
- Iterables can be iterated multiple times; iterators are single-use.
