## Method Overriding (Inheritance-Based)

```python
class Animal:
    """Base animal."""

    def speak(self):
        raise NotImplementedError


class Dog(Animal):
    def speak(self):
        return "Woof!"


class Cat(Animal):
    def speak(self):
        return "Meow!"


class Duck(Animal):
    def speak(self):
        return "Quack!"


def make_sound(animal):
    """Works with any Animal subclass."""
    print(animal.speak())


for a in [Dog(), Cat(), Duck()]:
    make_sound(a)
```

## Duck Typing

"If it walks like a duck and quacks like a duck..."

```python
class Duck:
    def quack(self):
        return "Quack!"


class Person:
    def quack(self):
        return "I'm pretending to be a duck!"


class Robot:
    def quack(self):
        return "BEEP QUACK BEEP"


def make_it_quack(thing):
    """Only checks that .quack() exists."""
    print(thing.quack())


make_it_quack(Duck())
make_it_quack(Person())
make_it_quack(Robot())
```

## Operator / Function Polymorphism

`len()` works on many different types.

```python
print(f"len('hello'): {len('hello')}")
print(f"len([1, 2, 3]): {len([1, 2, 3])}")
print(f"len({{'a': 1, 'b': 2}}): {len({'a': 1, 'b': 2})}")
print(f"len((1, 2, 3, 4)): {len((1, 2, 3, 4))}")
print(f"len({{1, 2, 3, 4, 5}}): {len({1, 2, 3, 4, 5})}")
```

## Protocol-Based Polymorphism

Python protocols are informal interfaces.

```python
class LoudList:
    """A list wrapper that announces length."""

    def __init__(self, items):
        self._items = items

    def __len__(self):
        print(f"Counting {len(self._items)} items...")
        return len(self._items)

    def __getitem__(self, index):
        return self._items[index]


loud = LoudList([10, 20, 30])
print(f"len: {len(loud)}")
print(f"loud[1]: {loud[1]}")
print(f"20 in loud: {20 in loud}")
```

## Key Takeaways

- Polymorphism: same interface, different behavior
- Method overriding + duck typing are both forms of polymorphism
- Built-in functions like `len()` are polymorphic
- Python protocols define informal interfaces via dunder methods
