## 1. Basic Unpacking

```python
# Tuple unpacking
point = (10, 20)
x, y = point
print(x, y)

# List unpacking
data = ["Alice", 30, "Engineer"]
name, age, job = data
print(name, age, job)
```

## 2. Swapping Variables

```python
a, b = 5, 10
a, b = b, a
print(a, b)  # 10 5
```

## 3. Extended Unpacking with *

```python
# Grab first, middle as list, last
nums = [1, 2, 3, 4, 5]
first, *middle, last = nums
print(first, middle, last)  # 1 [2, 3, 4] 5
```

```python
# Star at beginning
head, *tail = [1, 2, 3, 4]
print(head, tail)  # 1 [2, 3, 4]

# Star at end
*init, last = [1, 2, 3, 4]
print(init, last)  # [1, 2, 3] 4
```

## 4. Throwaway Variable _

```python
data = ("Alice", 30, "NYC", "555-1234")
name, _, city, _ = data
print(name, city)  # Alice NYC
```

## 5. Unpacking in For Loops

```python
students = [("Alice", 85), ("Bob", 92), ("Charlie", 78)]
for name, score in students:
    print(f"{name}: {score}")
```

## 6. Nested Unpacking

```python
nested = (1, (2, 3), 4)
a, (b, c), d = nested
print(a, b, c, d)  # 1 2 3 4
```

## Summary
- Unpack sequences into variables: `a, b = (1, 2)`
- Swap with `a, b = b, a`
- Use `*` for extended unpacking
- Use `_` for values you want to ignore
- Unpack in `for` loops and from function returns
