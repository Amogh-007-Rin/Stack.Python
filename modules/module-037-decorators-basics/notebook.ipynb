Enhance functions without changing their code.

## Functions as First-Class Objects

Functions can be assigned and passed around:

```python
def greet(name):
    return f"Hello, {name}!"

f = greet
print(f("Alice"))
print(type(f))
```

## Nested Functions (Closures)

Inner functions can access outer variables:

```python
def outer(msg):
    def inner(name):
        return f"{msg}, {name}!"
    return inner

hello = outer("Hello")
print(hello("Bob"))

hi = outer("Hi")
print(hi("Charlie"))
```

## Simple Decorator

A decorator wraps a function to add behavior:

```python
def logger(func):
    """Log when a function is called."""
    def wrapper(*args, **kwargs):
        print(f"Calling {func.__name__}...")
        result = func(*args, **kwargs)
        print(f"Finished {func.__name__}")
        return result
    return wrapper

@logger
def say_hello(name):
    """Greet someone."""
    print(f"Hello, {name}!")

say_hello("Alice")
```

## Timing Decorator

Measure how long a function takes:

```python
import time

def timer(func):
    """Print the execution time of a function."""
    def wrapper(*args, **kwargs):
        start = time.perf_counter()
        result = func(*args, **kwargs)
        elapsed = time.perf_counter() - start
        print(f"{func.__name__} took {elapsed:.4f}s")
        return result
    return wrapper

@timer
def slow_sum(n):
    """Sum numbers slowly."""
    total = 0
    for i in range(n):
        total += i
        time.sleep(0.0001)
    return total

result = slow_sum(100)
print(f"Result: {result}")
```

## Preserving Metadata with @wraps

Without @wraps, the function name and docstring are lost:

```python
from functools import wraps

def logger_with_wraps(func):
    """Logger decorator that preserves metadata."""
    @wraps(func)
    def wrapper(*args, **kwargs):
        """Wrapper docstring."""
        print(f"Calling {func.__name__}")
        return func(*args, **kwargs)
    return wrapper

@logger_with_wraps
def greet(name):
    """Greet someone nicely."""
    print(f"Hello, {name}!")

print(f"Name: {greet.__name__}")
print(f"Doc: {greet.__doc__}")
```

## Multiple Decorators

Decorators stack from bottom to top:

```python
@logger_with_wraps
@timer
def compute(n):
    """Compute sum of squares."""
    return sum(i * i for i in range(n))

result = compute(1000)
print(f"Result: {result}")
```

## Key Takeaways

- Decorators wrap functions to add behavior.
- `@decorator` is sugar for `func = decorator(func)`.
- Always use `@wraps` to preserve metadata.
- Decorators stack bottom-up.
- Use `*args, **kwargs` for flexible wrappers.
