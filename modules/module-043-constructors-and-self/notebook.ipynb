## Parameterized Constructors

Pass arguments to customize each instance.

```python
class Point:
    """A 2D point."""

    def __init__(self, x, y):
        self.x = x
        self.y = y

    def distance_from_origin(self):
        return (self.x ** 2 + self.y ** 2) ** 0.5


p1 = Point(3, 4)
p2 = Point(-1, 2)
print(f"p1: ({p1.x}, {p1.y}), distance: {p1.distance_from_origin():.2f}")
print(f"p2: ({p2.x}, {p2.y}), distance: {p2.distance_from_origin():.2f}")
```

## Default Values in `__init__`

```python
class Timer:
    """A simple countdown timer."""

    def __init__(self, duration=10, unit="seconds"):
        self.duration = duration
        self.unit = unit
        self.remaining = duration

    def __repr__(self):
        return f"Timer({self.duration}, {self.unit!r})"


t1 = Timer()
t2 = Timer(30)
t3 = Timer(60, "minutes")
print(t1)
print(t2)
print(t3)
```

## Mutable Default Trap

Never use mutable objects as default arguments!

```python
class Bad:
    """Bad example: mutable default."""
    def __init__(self, items=[]):
        self.items = items


class Good:
    """Good example: None + create new list."""
    def __init__(self, items=None):
        self.items = items if items is not None else []


a = Bad()
b = Bad()
a.items.append("shared!")
print(f"Bad: a={a.items}, b={b.items}")  # both have 'shared!'

c = Good()
d = Good()
c.items.append("unique")
print(f"Good: c={c.items}, d={d.items}")  # independent
```

## Python's Implicit self

Python passes the instance as the first argument automatically.

```python
class Demo:
    """Demonstrate implicit self."""

    def show(self, msg):
        print(f"Instance {id(self)}: {msg}")


d = Demo()
d.show("called via instance")         # Python adds d as self
Demo.show(d, "called via class")       # equivalent explicit call
```

## Key Takeaways

- `__init__` runs automatically when an object is created
- Parameterized constructors let you customize each instance
- Avoid mutable default arguments (use `None` instead)
- `self` is the instance itself, passed implicitly
