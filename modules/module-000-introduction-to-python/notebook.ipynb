### Key Principles

- **Beautiful is better than ugly.**
- **Explicit is better than implicit.**
- **Simple is better than complex.**
- **Complex is better than complicated.**
- **Readability counts.**
- **There should be one — and preferably only one — obvious way to do it.**

These aphorisms shape every design decision in the language.

## A Brief History

- **1980s**: Guido van Rossum begins work on Python during Christmas holidays.
- **1991**: First public release (Python 0.9.0).
- **1994**: Python 1.0 released.
- **2000**: Python 2.0 — list comprehensions, garbage collection.
- **2008**: Python 3.0 — major cleanup, breaking changes from Python 2.
- **2020**: Python 2 reaches end of life.
- **2023**: Python 3.12 is the latest stable version.

The name "Python" comes from *Monty Python's Flying Circus*, not the snake.

## How Python Runs Your Code

```
source.py  -->  Parser  -->  AST  -->  Compiler  -->  Bytecode (.pyc)  -->  Virtual Machine
```

1. **Parser** checks syntax.
2. **Compiler** translates to bytecode.
3. **Virtual Machine** executes bytecode.

This hybrid of compilation and interpretation gives Python portability and readability.

## The PEP Process

A **Python Enhancement Proposal (PEP)** is a design document for proposing changes.
Notable PEPs:
- **PEP 8**: Style Guide
- **PEP 20**: Zen of Python
- **PEP 572**: Walrus operator `:=`

Anyone can submit a PEP — it's an open process.

## Key Takeaways

- Created by Guido van Rossum, first released 1991.
- Named after Monty Python.
- Philosophy: readability, simplicity, explicit over implicit.
- Used everywhere: web, data, ML, automation, games, embedded.
- Source → bytecode → interpreter (CPython).
- PEPs guide language evolution.
- Always use Python 3 (never Python 2).
