# Module 014: Loops: `while`

Learn how to repeat code with `while` loops.

## 1. Basic `while` Loop

```python
count = 0
while count < 3:
    print(f"Count is {count}")
    count = count + 1
print("Loop finished.")
```

## 2. Counting with `while`

```python
i = 1
while i <= 5:
    print(i, end=" ")
    i = i + 1
print()
print(f"After loop: i = {i}")
```

## 3. Input Validation

```python
age = None
while age is None:
    user_input = input("Enter your age: ")
    if user_input.isdigit():
        age = int(user_input)
    else:
        print("Please enter a number.")
print(f"You are {age} years old.")
```

## 4. Sentinel-Controlled Loop

```python
total = 0
print("Enter numbers to sum. Enter 0 to stop.")
num = int(input("Enter a number: "))
while num != 0:
    total = total + num
    num = int(input("Enter a number: "))
print(f"Total: {total}")
```

## 5. Accumulator Pattern

```python
# Sum: start at 0
total = 0
i = 1
while i <= 100:
    total = total + i
    i = i + 1
print(f"Sum 1-100: {total}")
```

```python
# Count: start at 0
count = 0
i = 1
while i <= 100:
    if i % 2 == 0:
        count = count + 1
    i = i + 1
print(f"Even numbers 1-100: {count}")
```

```python
# Product: start at 1
product = 1
i = 1
while i <= 5:
    product = product * i
    i = i + 1
print(f"5! = {product}")
```

## 6. Number Guessing Game

```python
import random

secret = random.randint(1, 10)
guess = None
attempts = 0

while guess != secret:
    user_input = input("Guess (1-10): ")
    if user_input.isdigit():
        guess = int(user_input)
        attempts = attempts + 1
        if guess < secret:
            print("Too low!")
        elif guess > secret:
            print("Too high!")
    else:
        print("Enter a number!")

print(f"Correct! It took {attempts} tries.")
```

## 7. Menu Loop

```python
while True:
    print("\n1. Say Hello")
    print("2. Count to 3")
    print("3. Quit")
    choice = input("Choose: ")
    if choice == "1":
        print("Hello!")
    elif choice == "2":
        i = 1
        while i <= 3:
            print(i)
            i = i + 1
    elif choice == "3":
        print("Bye!")
        break
    else:
        print("Invalid.")
```
