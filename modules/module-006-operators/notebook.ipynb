## Arithmetic Operators

```python
a = 10
b = 3

print("a + b =", a + b)    # 13
print("a - b =", a - b)    # 7
print("a * b =", a * b)    # 30
print("a / b =", a / b)    # 3.333...
print("a // b =", a // b)  # 3 (floor division)
print("a % b =", a % b)    # 1 (remainder)
print("a ** b =", a ** b)  # 1000 (10 ^ 3)
```

### Floor Division with Negatives

```python
print("10 // 3 =", 10 // 3)    # 3
print("-10 // 3 =", -10 // 3)  # -4 (rounds DOWN)
```

### Modulo for Even/Odd Check

```python
num = 7
print(num, "is even?", num % 2 == 0)
print(num, "is odd?", num % 2 == 1)
```

## Comparison Operators

All comparison operators return `True` or `False`.

```python
x = 10
y = 5

print("x == y:", x == y)
print("x != y:", x != y)
print("x < y:", x < y)
print("x > y:", x > y)
print("x <= 10:", x <= 10)
print("x >= 10:", x >= 10)
print("Chained 18 <= x < 65:", 18 <= x < 65)
```

## Logical Operators

`and`, `or`, `not`

```python
age = 20
has_id = True

print("age >= 18 and has_id:", age >= 18 and has_id)
print("age >= 18 or has_id:", age >= 18 or has_id)
print("not has_id:", not has_id)
print("not age >= 18:", not age >= 18)
```

## Assignment Operators

Augmented assignment combines operation with assignment.

```python
count = 10
print("Initial:", count)

count += 5
print("After += 5:", count)

count -= 3
print("After -= 3:", count)

count *= 2
print("After *= 2:", count)

count //= 4
print("After //= 4:", count)
```

## Operator Precedence

Python follows `**` → `*`, `/`, `//`, `%` → `+`, `-` → comparison → `not` → `and` → `or`.

```python
result = 5 + 3 * 2 ** 3
print("5 + 3 * 2 ** 3 =", result)  # 29

result2 = (5 + 3) * 2 ** 3
print("(5 + 3) * 2 ** 3 =", result2)  # 64
```

## Number Analyzer Program

```python
num = 7

print("Square:", num ** 2)
print("Cube:", num ** 3)
print("Square root:", num ** 0.5)
print("Positive?", num > 0)
print("Negative?", num < 0)
print("Zero?", num == 0)
print("Even?", num % 2 == 0)
print("Between 1 and 100?", 1 <= num <= 100)

total = 0
total += num
total += 10
print("Total after adding num and 10:", total)
```

## Key Takeaways

- Arithmetic: `+`, `-`, `*`, `/`, `//`, `%`, `**`.
- Comparison returns `bool`: `==`, `!=`, `<`, `>`, `<=`, `>=`.
- Logical: `and` (both), `or` (either), `not` (negate).
- Augmented assignment: `+=`, `-=`, `*=`, `/=`, `//=`, `%=`, `**=`.
- Precedence: `**` > unary > `*/` > `+-` > comparisons > `not` > `and` > `or`.
- `//` rounds down (floor), not toward zero.
- `% 2` checks even/odd.
