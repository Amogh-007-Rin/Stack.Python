# Module 044: Class Variables vs Instance Variables

> **Phase:** 5. OOP  **Estimated time:** 1.5 hours

Shared state vs per-instance data.

## Class Variables

Defined outside methods, shared across all instances.

```python
class Employee:
    """Employee with class and instance variables."""

    company = "TechCorp"
    employee_count = 0

    def __init__(self, name, salary):
        self.name = name
        self.salary = salary
        Employee.employee_count += 1


e1 = Employee("Alice", 70000)
e2 = Employee("Bob", 80000)

print(f"Company: {Employee.company}")
print(f"e1.company: {e1.company}")
print(f"e2.company: {e2.company}")
print(f"Total employees: {Employee.employee_count}")
```

## Modifying Class vs Instance Variables

Assigning to `instance.var` creates a new instance variable, shadowing the class variable.

```python
class Employee:
    company = "TechCorp"


e1 = Employee()
e2 = Employee()

# Modify via class — affects all instances
Employee.company = "NewCorp"
print(f"e1.company after class change: {e1.company}")
print(f"e2.company after class change: {e2.company}")

# Assign to instance — creates instance var, shadows class var
e1.company = "PersonalCo"
print(f"e1.company after instance assign: {e1.company}")
print(f"e2.company unaffected: {e2.company}")
print(f"Employee.company: {Employee.company}")
```

## Mutable Class Variable Danger

Mutating a class variable through an instance affects all instances.

```python
class Team:
    members = []  # class variable (mutable!)

    def __init__(self, name):
        self.name = name

    def add_member(self, member):
        self.members.append(member)  # mutates the class-level list!


t1 = Team("Alpha")
t2 = Team("Beta")

t1.add_member("Alice")
print(f"t1.members: {t1.members}")
print(f"t2.members: {t2.members}")  # also has Alice!
print(f"Team.members: {Team.members}")
```

## `__dict__` for Attribute Inspection

```python
class Point:
    origin_name = "O"

    def __init__(self, x, y):
        self.x = x
        self.y = y


p = Point(3, 4)
print(f"Instance __dict__: {p.__dict__}")
print(f"Class __dict__ keys: {[k for k in Point.__dict__ if not k.startswith('_')]}")
```

## Key Takeaways

- Class variables are shared across all instances
- Instance variables are unique to each instance
- Assigning to `self.var` shadows class variables
- Be careful with mutable class variables
- `__dict__` shows all attributes of an object
