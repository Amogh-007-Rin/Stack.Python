Write anonymous inline functions with `lambda`.

## Lambda Syntax

`lambda arguments: expression` — a function without a name:

```python
square = lambda x: x ** 2
print(square(5))
print(square(10))
```

## Multiple Arguments

Lambdas can take any number of arguments:

```python
add = lambda a, b: a + b
print(add(3, 7))

max_val = lambda a, b: a if a > b else b
print(max_val(10, 20))
```

## Sorting with `key=lambda`

Custom sort keys are a common use for lambdas:

```python
students = [("Alice", 85), ("Bob", 72), ("Charlie", 90)]
students.sort(key=lambda s: s[1])
print(students)

words = ["python", "java", "c", "javascript"]
sorted_by_len = sorted(words, key=lambda w: len(w))
print(sorted_by_len)
```

## `map()` with Lambda

Transform every element in an iterable:

```python
nums = [1, 2, 3, 4, 5]
doubled = list(map(lambda x: x * 2, nums))
print(doubled)

celsius = [0, 20, 30, 100]
fahrenheit = list(map(lambda c: (c * 9/5) + 32, celsius))
print(fahrenheit)
```

## `filter()` with Lambda

Keep only elements where the predicate returns True:

```python
nums = [1, 2, 3, 4, 5, 6]
evens = list(filter(lambda x: x % 2 == 0, nums))
print(evens)

words = ["cat", "", "dog", "", "bird"]
non_empty = list(filter(lambda w: len(w) > 0, words))
print(non_empty)
```

## Lambda Limitations

Only a single expression — no statements allowed:

```python
# This works — single expression
is_even = lambda x: x % 2 == 0
print(is_even(4))

# This would NOT work — statements not allowed:
# lambda x: if x > 0: return x  # SyntaxError
```

## Lambda vs def

Use `def` for anything complex. Lambda for short, one-off functions:

```python
# Lambda — concise for simple operations
by_grade = lambda student: student[1]

# Equivalent def — better for complex logic
def by_grade_def(student):
    """Return the grade from a student tuple."""
    return student[1]

print("Both work the same way.")
```

## Key Takeaways

- `lambda args: expression` — anonymous inline function.
- Single expression only (no statements).
- Great for sorting keys, map, filter, short callbacks.
- Use `def` for anything complex or multi-line.
