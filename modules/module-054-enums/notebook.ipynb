# Module 054: Enums

```python
# Basic Enum
from enum import Enum


class Color(Enum):
    RED = 1
    GREEN = 2
    BLUE = 3


print(Color.RED)
print(Color.RED.name, Color.RED.value)
print(Color(1))
```

```python
# auto() for automatic values
from enum import auto


class Status(Enum):
    PENDING = auto()
    ACTIVE = auto()
    INACTIVE = auto()
    ARCHIVED = auto()


for status in Status:
    print(f"{status.name} = {status.value}")
```

```python
# IntEnum for integer-compatible enums
from enum import IntEnum


class Priority(IntEnum):
    LOW = 1
    MEDIUM = 5
    HIGH = 10


print(Priority.HIGH > Priority.LOW)
print(Priority(5))
```

```python
# StrEnum (Python 3.11+)
try:
    from enum import StrEnum

    class Direction(StrEnum):
        NORTH = "N"
        SOUTH = "S"
        EAST = "E"
        WEST = "W"

    print(Direction.NORTH.upper())
except ImportError:
    print("StrEnum requires Python 3.11+")
```
