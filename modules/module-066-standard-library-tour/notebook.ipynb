# Module 066: The Python Standard Library Tour (os, sys, datetime, random)

```python
# os module: working directory and listing
import os
print("Current directory:", os.getcwd())
print("Files here:", os.listdir(".")[:5])
```

```python
# os.path operations
paths = ["/home/user/docs/file.txt", "data/config.json", "./relative/path/"]
for p in paths:
    print(f"{p:40s} -> dir={os.path.dirname(p):20s} base={os.path.basename(p)}")
```

```python
# sys module: argv, version, path
import sys
print(f"Python version: {sys.version}")
print(f"First 3 sys.path entries:")
for p in sys.path[:3]:
    print(f"  {p}")
```

```python
# datetime module: now, timedelta, formatting
from datetime import datetime, timedelta

now = datetime.now()
future = now + timedelta(days=7, hours=3)
print(f"Now:   {now.strftime('%Y-%m-%d %H:%M:%S')}")
print(f"Future:{future.strftime('%Y-%m-%d %H:%M:%S')}")
print(f"Difference: {future - now}")
```

```python
# random module: random, randint, choice, shuffle, sample
import random

print(f"Random float: {random.random():.3f}")
print(f"Random int 1-10: {random.randint(1, 10)}")

fruits = ["apple", "banana", "cherry", "date", "elderberry"]
print(f"Random choice: {random.choice(fruits)}")
print(f"Random sample (3): {random.sample(fruits, 3)}")

random.shuffle(fruits)
print(f"Shuffled: {fruits}")
```
