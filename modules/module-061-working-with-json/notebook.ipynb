# Module 061: Working with JSON

```python
import json

# Serializing a dict to a JSON string
data = {"name": "Alice", "age": 30, "city": "New York"}
json_str = json.dumps(data)
print(json_str)
print(type(json_str))
```

```python
# Writing JSON to a file
with open("data.json", "w") as f:
    json.dump(data, f)
print("JSON written to data.json")
```

```python
# Parsing a JSON string into a Python dict
json_input = '{"product": "Laptop", "price": 1200.50, "in_stock": true}'
parsed = json.loads(json_input)
print(parsed)
print(parsed["price"])
```

```python
# Reading JSON from a file
with open("data.json", "r") as f:
    loaded = json.load(f)
print(loaded)
```

```python
# Pretty-printing with indent and sort_keys
nested = {
    "name": "Root",
    "children": [
        {"name": "Child1", "props": {"color": "red"}},
        {"name": "Child2", "props": {"color": "blue"}},
    ],
}
print(json.dumps(nested, indent=2, sort_keys=True))
```

```python
# JSON type mapping
python_obj = {
    "string": "hello",
    "int": 42,
    "float": 3.14,
    "bool": True,
    "none": None,
    "list": [1, 2, 3],
    "dict": {"key": "value"},
}
print(json.dumps(python_obj, indent=2))
```
