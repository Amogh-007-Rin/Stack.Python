## Defining a Simple Class

A class is a blueprint; an object is an instance built from that blueprint.

```python
class Dog:
    """A simple Dog class."""

    def __init__(self, name):
        self.name = name

    def bark(self):
        print(f"{self.name} says Woof!")


fido = Dog("Fido")
fido.bark()
```

## Creating Multiple Instances

Each instance has its own data.

```python
class Cat:
    """A simple Cat class."""

    def __init__(self, name, color):
        self.name = name
        self.color = color

    def meow(self):
        print(f"{self.color} cat {self.name} says Meow!")


cats = [
    Cat("Whiskers", "orange"),
    Cat("Shadow", "black"),
    Cat("Mittens", "white"),
]

for c in cats:
    c.meow()
```

## Class Without `__init__`

You can add attributes manually (but it's fragile).

```python
class Car:
    """Car class without __init__."""
    pass


my_car = Car()
my_car.make = "Toyota"
my_car.model = "Corolla"
my_car.year = 2020

print(f"{my_car.year} {my_car.make} {my_car.model}")

# But this fails:
# other = Car()
# print(other.make)  # AttributeError!
```

## Why OOP Matters

- **Encapsulation**: data + methods bundled together
- **Reusability**: one class, many instances
- **Organization**: groups related code logically
- **Modeling**: matches real-world entities directly

## Key Takeaways

- Class = blueprint; Object = instance
- `__init__` initializes new objects
- `self` refers to the current instance
- OOP helps organize and scale code
