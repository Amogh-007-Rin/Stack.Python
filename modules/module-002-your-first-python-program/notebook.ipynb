## The `print()` Function

`print()` displays text to the screen. The text must be inside quotes.

```python
print("Hello, World!")
```

### Single vs Double Quotes

Both work the same way. Just be consistent.

```python
print('Single quotes')
print("Double quotes")
```

## Multiple Arguments

Separate items with commas. Python adds a space between them.

```python
print("Python", "is", "fun")
```

## Simple Arithmetic in `print()`

Python evaluates math expressions before printing the result.

```python
print(2 + 3)
print(10 - 4)
print(3 * 7)
print(8 / 2)
```

Some more examples combining text and numbers:

```python
print("Two plus three is", 2 + 3)
print("Ten minus four is", 10 - 4)
```

## Comments

Comments start with `#`. Python ignores everything after `#` on that line.

```python
# This is a comment
print("This code runs")  # This part is also a comment
# print("This won't run because it's commented out")
```

## Run a .py File

In a real terminal, you'd run `python3 filename.py`. In this notebook, each cell acts like a mini-script.

```python
print("This is the first line of the script")
print("This is the second line")
# A comment between lines
print(100 - 99)
```

## Key Takeaways

- `print()` displays output to the screen.
- Strings need matching quotes (single or double).
- Commas in `print()` add spaces between arguments.
- Math inside `print()` is evaluated first.
- Comments (`#`) are for humans, ignored by Python.
- Division (`/`) always returns a float.
- Always use parentheses with `print()` in Python 3.
