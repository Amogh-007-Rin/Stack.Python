## Factorial — The Classic Example

Base case: n <= 1. Recursive case: n * factorial(n-1):

```python
def factorial(n: int) -> int:
    """Calculate n! recursively.

    Args:
        n: A non-negative integer

    Returns:
        n factorial
    """
    if n <= 1:
        return 1
    return n * factorial(n - 1)

for i in range(6):
    print(f"{i}! = {factorial(i)}")
```

## Fibonacci

Base case: n <= 1. Recursive case: fib(n-1) + fib(n-2):

```python
def fibonacci(n: int) -> int:
    """Return the nth Fibonacci number.

    Args:
        n: Position in Fibonacci sequence

    Returns:
        The nth Fibonacci number
    """
    if n <= 1:
        return n
    return fibonacci(n - 1) + fibonacci(n - 2)

for i in range(10):
    print(f"fib({i}) = {fibonacci(i)}")
```

## Recursion Tree Visualization

See how fibonacci(5) branches:

```python
def fib_with_trace(n: int, depth: int = 0) -> int:
    """Fibonacci with indentation to show the recursion tree.

    Args:
        n: Position in Fibonacci sequence
        depth: Current recursion depth (for indentation)

    Returns:
        The nth Fibonacci number
    """
    indent = "  " * depth
    print(f"{indent}fib({n})")
    if n <= 1:
        return n
    result = fib_with_trace(n - 1, depth + 1) + fib_with_trace(n - 2, depth + 1)
    print(f"{indent}=> {result}")
    return result

print("\nRecursion tree for fib(5):")
fib_with_trace(5)
```

## Countdown

A simple recursive countdown:

```python
def countdown(n: int):
    """Print countdown from n to 1.

    Args:
        n: Starting number
    """
    if n <= 0:
        print("Liftoff!")
        return
    print(n)
    countdown(n - 1)

countdown(5)
```

## Stack Depth Limit

Python limits recursion depth to prevent stack overflow:

```python
import sys
print(f"Recursion limit: {sys.getrecursionlimit()}")

def recurse(n: int):
    """Recurse until RecursionError."""
    print(n, end=" ")
    recurse(n + 1)

try:
    recurse(1)
except RecursionError:
    print("\nRecursionError: maximum recursion depth exceeded")
```

## Iteration vs Recursion

Same problem solved both ways:

```python
def factorial_iterative(n: int) -> int:
    """Calculate n! iteratively.

    Args:
        n: A non-negative integer

    Returns:
        n factorial
    """
    result = 1
    for i in range(2, n + 1):
        result *= i
    return result

def factorial_recursive(n: int) -> int:
    """Calculate n! recursively.

    Args:
        n: A non-negative integer

    Returns:
        n factorial
    """
    if n <= 1:
        return 1
    return n * factorial_recursive(n - 1)

print(f"Iterative: {factorial_iterative(10)}")
print(f"Recursive: {factorial_recursive(10)}")
```

## Key Takeaways

- Recursion = base case + recursive case that moves toward the base case.
- Every recursive call adds a frame to the call stack.
- Python has a recursion limit (~1000).
- Use iteration for deep recursion; recursion for self-similar structures.
