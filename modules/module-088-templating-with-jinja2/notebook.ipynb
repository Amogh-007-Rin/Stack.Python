## Jinja2 Template Basics

```python
from jinja2 import Template

# Simple template
tpl: Template = Template('Hello, {{ name }}!')
print(tpl.render(name='World'))

# With filters
tpl2: Template = Template('{{ items|length }} items: {{ items|join(", ") }}')
print(tpl2.render(items=['apple', 'banana', 'cherry']))
```

## Loops and Conditionals

```python
tpl_loop: str = '''
<ul>
{% for item in items %}
    <li>{{ item|capitalize }}</li>
{% endfor %}
</ul>
'''
t: Template = Template(tpl_loop)
print(t.render(items=['apple', 'banana', 'cherry']))
```

```python
tpl_if: str = '''
{% if score >= 90 %}
    Excellent!
{% elif score >= 70 %}
    Good
{% else %}
    Needs Improvement
{% endif %}
'''
t = Template(tpl_if)
for s in [95, 75, 50]:
    print(f'Score {s}: {t.render(score=s).strip()}')
```

## Template Inheritance

```python
base_template: str = '''
<html>
<head><title>{% block title %}My Site{% endblock %}</title></head>
<body>
  <nav><a href="/">Home</a> | <a href="/about">About</a></nav>
  {% block content %}{% endblock %}
</body>
</html>
'''

from jinja2 import Environment, BaseLoader

env = Environment(loader=BaseLoader())

# Register base template
base = env.from_string(base_template)
env.globals['base_template'] = base_template

home_page: str = '''
{% extends base_template %}
{% block title %}Home{% endblock %}
{% block content %}<h1>Welcome!</h1>{% endblock %}
'''
t = env.from_string(home_page)
print(t.render(base_template=base_template))
```

## Flask Integration with render_template

```python
from flask import Flask, render_template

app: Flask = Flask(__name__)

# In a real app, templates are stored in templates/ folder
# Here we demonstrate the pattern:

@app.route('/')
def home() -> str:
    return render_template('home.html', title='Home', items=['Task 1', 'Task 2', 'Task 3'])

# Test client
with app.test_request_context('/'):
    print('Route registered: /')
    print('Flask + Jinja2 ready for template rendering')
```

## Creating Templates Programmatically

```python
import tempfile
import os

# Create a simple template file to demonstrate
tmpdir: str = tempfile.mkdtemp()
tpl_path: str = os.path.join(tmpdir, 'hello.html')
with open(tpl_path, 'w') as f:
    f.write('<h1>Hello, {{ name }}!</h1>')

from jinja2 import FileSystemLoader, Environment
loader = FileSystemLoader(tmpdir)
env = Environment(loader=loader)
tpl = env.get_template('hello.html')
print(tpl.render(name='World'))

# Cleanup
os.remove(tpl_path)
os.rmdir(tmpdir)
```
