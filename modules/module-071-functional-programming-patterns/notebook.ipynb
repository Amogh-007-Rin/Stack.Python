## Pure Functions

A pure function has no side effects and always returns the same output for the same input.

```python
# Pure function
def add_one(x: int) -> int:
    return x + 1

# Impure function (has side effect)
result_cache = []
def impure_add_one(x: int) -> int:
    result_cache.append(x)
    return x + 1
```

## map, filter, reduce

```python
from functools import reduce

numbers = [1, 2, 3, 4, 5]

# map
squared = list(map(lambda x: x ** 2, numbers))
print('map:', squared)

# filter
evens = list(filter(lambda x: x % 2 == 0, numbers))
print('filter:', evens)

# reduce
total = reduce(lambda acc, x: acc + x, numbers, 0)
print('reduce:', total)
```

## functools.partial

```python
from functools import partial

def power(base: int, exp: int) -> int:
    return base ** exp

square = partial(power, exp=2)
cube = partial(power, exp=3)

print('square of 5:', square(5))
print('cube of 3:', cube(3))
```

## Operator Module

```python
from operator import add, itemgetter, attrgetter

# add
print('add(3, 4):', add(3, 4))

# itemgetter
data = [{'name': 'Alice', 'age': 30}, {'name': 'Bob', 'age': 25}]
get_age = itemgetter('age')
print('ages:', list(map(get_age, data)))

# sorted with itemgetter
sorted_data = sorted(data, key=itemgetter('age'))
print('sorted:', sorted_data)
```

## functools.lru_cache

```python
from functools import lru_cache

@lru_cache(maxsize=None)
def fib(n: int) -> int:
    if n < 2:
        return n
    return fib(n - 1) + fib(n - 2)

print('fib(50):', fib(50))
```
