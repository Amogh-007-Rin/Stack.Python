Functions that take or return other functions.

## Functions as First-Class Objects

Functions can be assigned, passed, and returned:

```python
def square(x):
    return x * x

f = square
print(f(5))

def apply(func, values):
    """Apply a function to each value in a list.

    Args:
        func: Function to apply
        values: List of values

    Returns:
        List of transformed values
    """
    return [func(v) for v in values]

print(apply(square, [1, 2, 3]))
```

## map() — Transform Each Element

Apply a function to every element:

```python
nums = [1, 2, 3, 4]

doubled = list(map(lambda x: x * 2, nums))
print(doubled)

words = ["hello", "world", "python"]
uppered = list(map(str.upper, words))
print(uppered)
```

## filter() — Select by Predicate

Keep elements where the function returns True:

```python
nums = [1, 2, 3, 4, 5, 6]
evens = list(filter(lambda x: x % 2 == 0, nums))
print(evens)

mixed = [0, None, "hello", "", [], [1, 2]]
truthy = list(filter(None, mixed))
print(truthy)
```

## reduce() — Accumulate

Cumulatively apply a function to reduce an iterable to a single value:

```python
from functools import reduce

total = reduce(lambda a, b: a + b, [1, 2, 3, 4])
print(total)

product = reduce(lambda a, b: a * b, [1, 2, 3, 4, 5])
print(product)

maximum = reduce(lambda a, b: a if a > b else b, [3, 7, 2, 9, 1])
print(maximum)
```

## sorted() with Key

Custom sorting with key functions:

```python
words = ["python", "java", "c", "javascript"]
print(sorted(words, key=len))

students = [("Alice", 85), ("Bob", 72), ("Charlie", 90)]
print(sorted(students, key=lambda s: s[1], reverse=True))
```

## Returning Functions

Functions that create and return other functions:

```python
def make_multiplier(n):
    """Return a function that multiplies by n.

    Args:
        n: The multiplier

    Returns:
        A function that takes x and returns x * n
    """
    def multiplier(x):
        return x * n
    return multiplier

double = make_multiplier(2)
triple = make_multiplier(3)

print(double(10))
print(triple(10))
```

## Key Takeaways

- `map()` transforms; `filter()` selects; `reduce()` accumulates.
- Functions are first-class: pass, return, assign.
- Use `sorted()` with `key=` for custom sorting.
- `map()`/`filter()` return iterators — call `list()` to materialize.
