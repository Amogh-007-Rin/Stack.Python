Learn about sets — unordered collections of unique elements.

## 1. Creating Sets

```python
# Using curly braces
fruits = {"apple", "banana", "cherry"}
print(fruits)

# Using set() constructor
nums = set([1, 2, 3, 2, 1])
print(nums)  # duplicates removed
```

```python
# Empty set requires set(), not {}
empty_set = set()
empty_dict = {}
print(type(empty_set), type(empty_dict))
```

## 2. Set Methods: Add, Remove, Discard

```python
s = {1, 2, 3}
s.add(4)
s.discard(2)     # safe removal
s.remove(1)      # raises KeyError if missing
print(s)
```

```python
# pop, clear, copy
s = {10, 20, 30}
val = s.pop()
print("Popped:", val, "Remaining:", s)
s.clear()
print("After clear:", s)
```

## 3. Set Operations

```python
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}

print("Union:", A | B)                # {1,2,3,4,5,6}
print("Intersection:", A & B)         # {3,4}
print("Difference A-B:", A - B)       # {1,2}
print("Symmetric diff:", A ^ B)       # {1,2,5,6}
```

## 4. Set Comprehensions

```python
evens = {x for x in range(20) if x % 2 == 0}
print(evens)
```

## 5. Frozenset

```python
fs = frozenset([1, 2, 3, 2, 1])
print(fs)
# fs.add(4)  # AttributeError — frozenset is immutable
```

## 6. Removing Duplicates

```python
names = ["Alice", "Bob", "Alice", "Charlie", "Bob"]
unique = list(set(names))
print(sorted(unique))
```

## Summary
- Sets store unique, unordered elements
- Use `{}` for non-empty, `set()` for empty
- Powerful set operations: `|`, `&`, `-`, `^`
- `frozenset` is the immutable version
- Great for deduplication and membership tests
