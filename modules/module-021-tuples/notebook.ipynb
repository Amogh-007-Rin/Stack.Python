Learn about tuples — immutable sequences in Python.

## 1. Creating Tuples

```python
# Using parentheses
empty = ()
point = (3, 4)
colors = ("red", "green", "blue")
print(empty, point, colors)
```

```python
# Using tuple() constructor
nums = tuple([1, 2, 3])
chars = tuple("abc")
print(nums, chars)
```

```python
# Single-element tuple needs trailing comma
single = (5,)
not_tuple = (5)
print(type(single), type(not_tuple))
```

## 2. Immutability vs Lists

```python
# Tuples cannot be changed after creation
t = (1, 2, 3)
try:
    t[0] = 99
except TypeError as e:
    print(f"Error: {e}")
```

```python
# Lists are mutable
lst = [1, 2, 3]
lst[0] = 99
print(lst)  # [99, 2, 3]
```

## 3. Indexing and Slicing

```python
t = (10, 20, 30, 40, 50)
print(t[0], t[-1], t[1:4], t[::-1])
```

## 4. When to Use Tuples

```python
# Returning multiple values from a function
def min_max(nums):
    return min(nums), max(nums)

result = min_max([3, 1, 7, 2, 9])
print(result)  # (1, 9)
```

```python
# Tuples as dictionary keys (lists cannot be used)
locations = {
    (0, 0): "Origin",
    (1, 2): "School",
}
print(locations[(0, 0)])
```

## 5. Tuple Unpacking

```python
# Basic unpacking
point = (4, 7)
x, y = point
print(f"x={x}, y={y}")

# Swapping variables
a, b = 5, 10
a, b = b, a
print(f"a={a}, b={b}")
```

```python
# Unpacking with *
first, *middle, last = (1, 2, 3, 4, 5)
print(first, middle, last)
```

## 6. Namedtuple

```python
from collections import namedtuple

Point = namedtuple("Point", ["x", "y"])
p = Point(3, 4)
print(p.x, p.y, p[0], p[1])
```

## Summary
- Tuples are immutable sequences created with `()`
- Use them for fixed data, dict keys, and multiple return values
- Tuple unpacking enables clean assignment
- `namedtuple` adds named fields to tuples
