# Module 003: Variables and Data Types

> **Phase:** 1. Fundamentals  **Estimated time:** 2 hours

Learn how to store data in variables and understand Python's basic data types.

## Variable Assignment

Use `=` to store a value under a name.

```python
name = "Alice"
age = 25
print(name)
print(age)
```

## Using Variables in Expressions

Once assigned, variables behave like the values they hold.

```python
x = 10
y = 5
print(x + y)
print(x - y)
print(x * y)
print(x / y)
```

## Dynamic Typing

A variable's type can change as the program runs.

```python
value = 42
print(value, type(value))
value = "now a string"
print(value, type(value))
value = 3.14
print(value, type(value))
```

## Basic Data Types

### int — Integer

```python
count = 10
negative = -5
large = 1_000_000
print(count, negative, large)
print(type(count))
```

### float — Floating-point number

```python
pi = 3.14159
temp = -2.5
sci = 1.5e10
print(pi, temp, sci)
print(type(pi))
```

### str — String

```python
greeting = "Hello"
name = 'Alice'
print(greeting, name)
print(type(greeting))
```

### bool — Boolean

```python
is_sunny = True
is_raining = False
print(is_sunny, is_raining)
print(type(is_sunny))
```

## The `type()` Function

Use `type()` to check what type a value is.

```python
print(type(42))
print(type(3.14))
print(type("Text"))
print(type(False))
```

## Naming Rules

- Start with letter or underscore
- Can contain letters, digits, underscores
- Case-sensitive: `Age` and `age` are different
- Cannot use reserved keywords (`if`, `for`, `while`, etc.)

```python
# Valid names
user_name = "Bob"
_count = 10
data2 = [1, 2, 3]
camelCase = "okay but not Pythonic"
snake_case = "preferred in Python"

print(user_name, _count, data2, camelCase, snake_case)
```

## Key Takeaways

- `=` assigns a value to a variable.
- Names: letter/underscore first, case-sensitive, no keywords.
- Dynamic typing lets variables change type.
- Core types: `int`, `float`, `str`, `bool`.
- `type()` tells you a value's data type.
- `True` and `False` are capitalized.
