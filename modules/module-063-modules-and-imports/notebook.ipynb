# Module 063: Modules and Imports

```python
# Different import styles
import math
from math import sqrt
import math as m

print(math.sqrt(16))
print(sqrt(16))
print(m.sqrt(16))
```

```python
# The __name__ == "__main__" idiom
print(f"__name__ is: {__name__}")
if __name__ == "__main__":
    print("This block runs only when executed directly")
```

```python
# Module search path
import sys
print("Python searches for modules in:")
for path in sys.path:
    print(f"  {path}")
```

```python
# Creating and importing a custom module
# First, write a module file
module_code = '''
def greet(name: str) -> str:
    return f"Hello, {name}!"

PI = 3.14159
'''

with open("mymodule.py", "w") as f:
    f.write(module_code)

# Now import it
import mymodule  # type: ignore
print(mymodule.greet("World"))
print(mymodule.PI)
```

```python
# Reloading modules
import importlib

# Modify the module file
with open("mymodule.py", "w") as f:
    f.write('''
def greet(name: str) -> str:
    return f"Hi there, {name}!"

PI = 3.14
''')

importlib.reload(mymodule)  # noqa: F821
print(mymodule.greet("World"))  # noqa: F821
print(mymodule.PI)  # noqa: F821
```

```python
import os
os.remove("mymodule.py")
print("Cleanup done")
```
