```python
# @classmethod: alternative constructors
from typing import Any


class Person:
    def __init__(self, first: str, last: str, age: int) -> None:
        self.first = first
        self.last = last
        self.age = age

    @classmethod
    def from_birth_year(cls, first: str, last: str, birth_year: int) -> "Person":
        return cls(first, last, 2025 - birth_year)

    @classmethod
    def from_dict(cls, data: dict[str, Any]) -> "Person":
        return cls(data["first"], data["last"], data["age"])


p1 = Person.from_birth_year("Alice", "Smith", 1990)
p2 = Person.from_dict({"first": "Bob", "last": "Jones", "age": 30})
print(p1.age, p2.first)
```

```python
# @staticmethod: utility inside class namespace
import math


class Geometry:
    @staticmethod
    def circle_area(radius: float) -> float:
        return math.pi * radius ** 2

    @staticmethod
    def rectangle_area(width: float, height: float) -> float:
        return width * height


print(Geometry.circle_area(5))
print(Geometry.rectangle_area(3, 4))
```

```python
# @property: computed attributes with getter/setter/deleter
class Temperature:
    def __init__(self, celsius: float = 0) -> None:
        self._celsius = celsius

    @property
    def celsius(self) -> float:
        return self._celsius

    @celsius.setter
    def celsius(self, value: float) -> None:
        if value < -273.15:
            raise ValueError("Below absolute zero")
        self._celsius = value

    @property
    def fahrenheit(self) -> float:
        return self._celsius * 9 / 5 + 32

    @fahrenheit.setter
    def fahrenheit(self, value: float) -> None:
        self._celsius = (value - 32) * 5 / 9

    @celsius.deleter
    def celsius(self) -> None:
        print("Deleting temperature")
        self._celsius = 0


t = Temperature(25)
print(t.fahrenheit)
t.fahrenheit = 100
print(t.celsius)
```
