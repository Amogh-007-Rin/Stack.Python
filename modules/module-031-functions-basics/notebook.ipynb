Learn how to define and call functions in Python.

## Defining a Function

Use `def`, a name, parentheses, and a colon:

```python
def greet():
    """Print a greeting message."""
    print("Hello, world!")

greet()
```

## Parameters and Arguments

A parameter is the placeholder; an argument is the actual value:

```python
def greet_person(name):
    """Greet a person by name.

    Args:
        name: The person's name
    """
    print(f"Hello, {name}!")

greet_person("Alice")
greet_person("Bob")
```

## The `return` Statement

Send a value back to the caller:

```python
def square(x):
    """Return the square of x.

    Args:
        x: A number

    Returns:
        x squared
    """
    return x * x

result = square(5)
print(f"5 squared = {result}")
```

## Multiple Parameters

Functions can take any number of parameters:

```python
def add(a: int, b: int) -> int:
    """Add two numbers and return the result.

    Args:
        a: First number
        b: Second number

    Returns:
        Sum of a and b
    """
    return a + b

print(add(3, 7))  # 10
print(add(-1, 5)) # 4
```

## Functions Without `return`

A function without `return` implicitly returns `None`:

```python
def just_print():
    """This function prints but does not return."""
    print("I print something")

result = just_print()
print(f"Return value: {result}")  # None
```

## Local Scope

Variables created inside a function are local:

```python
def my_func():
    local_var = "I'm local"
    print(local_var)

my_func()

try:
    print(local_var)
except NameError as e:
    print(f"Error: {e}")
```

## Function Naming

Use `snake_case`, descriptive verbs, and follow PEP 8:

```python
def calculate_average(numbers):
    """Calculate the average of a list of numbers.

    Args:
        numbers: List of numeric values

    Returns:
        The arithmetic mean
    """
    return sum(numbers) / len(numbers)

print(calculate_average([10, 20, 30]))
```

## Key Takeaways

- `def` defines a function; `return` sends a result back.
- Parameters are placeholders; arguments are actual values.
- Variables inside functions are local by default.
- Use Google-style docstrings for all functions.
- Name functions with `snake_case`.
