## shutil: Copy, Move, Archive

```python
import shutil
import os
from pathlib import Path

# Setup test directory
test_dir: Path = Path('test_automation')
test_dir.mkdir(exist_ok=True)
(test_dir / 'file1.txt').write_text('Hello')
(test_dir / 'file2.txt').write_text('World')

# Copy
shutil.copy(test_dir / 'file1.txt', 'file1_copy.txt')
print(f'Copy exists: {os.path.exists("file1_copy.txt")}')

# Move
shutil.move('file1_copy.txt', 'file1_moved.txt')
print(f'Move: source gone={not os.path.exists("file1_copy.txt")}, dest exists={os.path.exists("file1_moved.txt")}')

# Archive
archive_path: str = shutil.make_archive('test_backup', 'zip', str(test_dir))
print(f'Archive: {archive_path} ({os.path.getsize(archive_path)} bytes)')

# Cleanup
os.remove('file1_moved.txt')
os.remove(archive_path)
shutil.rmtree(test_dir)
print('Cleanup done')
```

## glob: File Pattern Matching

```python
import glob
from typing import List

# Create sample files for demonstration
for name in ['doc1.txt', 'doc2.txt', 'image1.png', 'script.py']:
    Path(name).write_text('sample')

txt_files: List[str] = glob.glob('*.txt')
print(f'Text files: {txt_files}')

all_files: List[str] = glob.glob('*.*')
print(f'All files: {all_files}')

# Recursive glob
py_files: List[str] = glob.glob('**/*.py', recursive=True)
print(f'Python files: {py_files}')

# Cleanup
for f in ['doc1.txt', 'doc2.txt', 'image1.png', 'script.py']:
    os.remove(f)
```

## os.walk: Directory Traversal

```python
# Create test structure
os.makedirs('demo/sub1', exist_ok=True)
os.makedirs('demo/sub2', exist_ok=True)
Path('demo/file_a.txt').write_text('a')
Path('demo/sub1/file_b.txt').write_text('b')
Path('demo/sub2/file_c.txt').write_text('c')

for dirpath, dirnames, filenames in os.walk('demo'):
    print(f'Directory: {dirpath}')
    for d in dirnames:
        print(f'  Subdir: {d}')
    for f in filenames:
        filepath: str = os.path.join(dirpath, f)
        print(f'  File: {f} ({os.path.getsize(filepath)} bytes)')
    print()

# Cleanup
shutil.rmtree('demo')
```

## Simple Scheduling

```python
import schedule
import time

job_run_count: int = 0

def sample_job() -> None:
    global job_run_count
    job_run_count += 1
    print(f'[{time.strftime("%H:%M:%S")}] Job executed (#{job_run_count})')

# Schedule a job every 2 seconds for demo
schedule.every(2).seconds.do(sample_job)

# Run for 5 seconds
start: float = time.time()
while time.time() - start < 5:
    schedule.run_pending()
    time.sleep(0.5)

print(f'\nJob ran {job_run_count} times')
schedule.clear()
```

## File Organizer Pattern

```python
EXTENSION_MAP: dict = {
    '.txt': 'Documents', '.pdf': 'Documents',
    '.jpg': 'Images', '.png': 'Images',
    '.mp3': 'Audio',
    '.zip': 'Archives',
}

def organize(path: str) -> None:
    for f in os.listdir(path):
        full = os.path.join(path, f)
        if os.path.isdir(full):
            continue
        ext = os.path.splitext(f)[1].lower()
        category = EXTENSION_MAP.get(ext, 'Others')
        cat_dir = os.path.join(path, category)
        os.makedirs(cat_dir, exist_ok=True)
        shutil.move(full, os.path.join(cat_dir, f))
        print(f'Moved {f} -> {category}/')

# Demo with temp dir
demo_path = 'organize_demo'
os.makedirs(demo_path, exist_ok=True)
for name in ['report.pdf', 'photo.jpg', 'song.mp3', 'notes.txt', 'data.zip']:
    Path(os.path.join(demo_path, name)).write_text('x')

organize(demo_path)
shutil.rmtree(demo_path)
```
