## Handling Missing Data

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'A': [1, 2, np.nan, 4],
    'B': [5, np.nan, np.nan, 8],
    'C': [9, 10, 11, 12]
})

print('Original:')
print(df)
print('\nNull counts:')
print(df.isnull().sum())
print('\nDrop rows with any null:')
print(df.dropna())
print('\nFill with 0:')
print(df.fillna(0))
```

## Duplicate Removal

```python
import pandas as pd

df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Alice', 'Charlie', 'Bob'],
    'Age': [25, 30, 25, 35, 30]
})

print('Duplicated rows:', df.duplicated().sum())
df_dedup = df.drop_duplicates()
print('After dedup:')
print(df_dedup)
```

## String Operations

```python
import pandas as pd

df = pd.DataFrame({
    'Name': ['Alice Smith', 'Bob Jones', 'Charlie Brown']
})

df['First'] = df['Name'].str.split().str[0]
df['Last'] = df['Name'].str.split().str[1]
df['Email'] = df['First'].str.lower() + '@company.com'
print(df)
```

## GroupBy and Aggregation

```python
import pandas as pd

df = pd.DataFrame({
    'Department': ['Sales', 'Sales', 'Engineering', 'Engineering'],
    'Employee': ['Alice', 'Bob', 'Charlie', 'Diana'],
    'Salary': [70000, 80000, 90000, 85000]
})

dept_stats = df.groupby('Department')['Salary'].agg(['mean', 'sum', 'count'])
print(dept_stats)
```

## Merging DataFrames

```python
import pandas as pd

employees = pd.DataFrame({
    'emp_id': [1, 2, 3],
    'name': ['Alice', 'Bob', 'Charlie'],
    'dept_id': [101, 102, 101]
})

departments = pd.DataFrame({
    'dept_id': [101, 102],
    'dept_name': ['Sales', 'Engineering']
})

merged = pd.merge(employees, departments, on='dept_id', how='left')
print(merged)
```

## Pivot Tables

```python
import pandas as pd

df = pd.DataFrame({
    'Year': [2022, 2022, 2023, 2023],
    'Region': ['East', 'West', 'East', 'West'],
    'Sales': [100, 150, 120, 170]
})

pivot = pd.pivot_table(
    df,
    values='Sales',
    index='Region',
    columns='Year',
    aggfunc='sum'
)
print(pivot)
```
