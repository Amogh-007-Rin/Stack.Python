## Creating an ABC

Use `ABC` and `@abstractmethod` to define required methods.

```python
from abc import ABC, abstractmethod


class Shape(ABC):
    """Abstract base class for shapes."""

    @abstractmethod
    def area(self):
        """Calculate area."""
        ...

    @abstractmethod
    def perimeter(self):
        """Calculate perimeter."""
        ...


class Circle(Shape):
    """Concrete circle."""

    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14159 * self.radius ** 2

    def perimeter(self):
        return 2 * 3.14159 * self.radius


class Rectangle(Shape):
    """Concrete rectangle."""

    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

    def perimeter(self):
        return 2 * (self.width + self.height)


shapes = [Circle(5), Rectangle(3, 4)]
for s in shapes:
    print(f"{type(s).__name__}: area={s.area():.2f}, perimeter={s.perimeter():.2f}")
```

## Cannot Instantiate ABC Directly

```python
from abc import ABC, abstractmethod


class Abstract(ABC):
    @abstractmethod
    def must_implement(self):
        ...


try:
    obj = Abstract()
except TypeError as e:
    print(f"Cannot instantiate ABC: {e}")

# Also cannot instantiate incomplete subclass:
# class Incomplete(Abstract):
#     pass
# obj = Incomplete()  # TypeError!
```

## Abstract Properties

```python
from abc import ABC, abstractmethod


class Employee(ABC):
    """Abstract employee."""

    @property
    @abstractmethod
    def role(self):
        ...

    @abstractmethod
    def calculate_pay(self):
        ...


class Manager(Employee):
    @property
    def role(self):
        return "Manager"

    def calculate_pay(self):
        return 80000


class Developer(Employee):
    @property
    def role(self):
        return "Developer"

    def calculate_pay(self):
        return 90000


for emp in [Manager(), Developer()]:
    print(f"{emp.role}: ${emp.calculate_pay()}")
```

## Registering Virtual Subclasses

```python
from abc import ABC


class MyIterable(ABC):
    @abstractmethod
    def __iter__(self):
        ...


# Register existing types as virtual subclasses
MyIterable.register(list)
MyIterable.register(tuple)
MyIterable.register(str)

print(f"[] is MyIterable: {isinstance([], MyIterable)}")
print(f"() is MyIterable: {isinstance((), MyIterable)}")
print(f"'hello' is MyIterable: {isinstance('hello', MyIterable)}")
print(f"42 is MyIterable: {isinstance(42, MyIterable)}")
```

## Key Takeaways

- ABCs define interfaces that subclasses must implement
- `@abstractmethod` marks required methods
- Concrete subclasses must implement all abstract methods
- Abstract properties enforce property contracts
- `.register()` creates virtual subclass relationships
- `collections.abc` has ready-made ABCs for common protocols
