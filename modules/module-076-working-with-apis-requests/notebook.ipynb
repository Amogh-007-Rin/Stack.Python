## Basic GET Request

```python
import requests

response = requests.get('https://jsonplaceholder.typicode.com/posts/1')
print('Status:', response.status_code)
print('Content-Type:', response.headers.get('Content-Type'))
data = response.json()
print('Title:', data['title'])
```

## Query Parameters and Headers

```python
import requests

# Query parameters
params = {'userId': 1}
response = requests.get('https://jsonplaceholder.typicode.com/posts', params=params)
print('Posts by user 1:', len(response.json()))

# Custom headers
headers = {'User-Agent': 'MyLearningApp/1.0'}
response = requests.get('https://httpbin.org/headers', headers=headers)
print('Headers response:', response.json())
```

## POST Request with JSON

```python
import requests

new_data = {
    'title': 'Learning APIs',
    'body': 'Python requests library is great!',
    'userId': 1
}
response = requests.post('https://jsonplaceholder.typicode.com/posts', json=new_data)
print('Created:', response.json())
```

## Error Handling

```python
import requests
from typing import Optional

def safe_get(url: str, timeout: int = 10) -> Optional[dict]:
    try:
        response = requests.get(url, timeout=timeout)
        response.raise_for_status()
        return response.json()
    except requests.exceptions.Timeout:
        print('Request timed out')
        return None
    except requests.exceptions.HTTPError as e:
        print(f'HTTP error: {e}')
        return None
    except requests.exceptions.RequestException as e:
        print(f'Request failed: {e}')
        return None

result = safe_get('https://jsonplaceholder.typicode.com/posts/1')
print('Result:', result['title'] if result else 'Failed')
```

## Authentication

```python
import requests

# API Key in header
headers = {'Authorization': 'Bearer test-token-123'}
response = requests.get('https://httpbin.org/bearer', headers=headers)
print('Auth response:', response.json())

# Using Session for reusability
session = requests.Session()
session.headers.update({'Authorization': 'Bearer test-token-123'})
response = session.get('https://httpbin.org/bearer')
print('Session response:', response.json())
```
