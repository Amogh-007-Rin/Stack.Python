Build a CSV-backed expense tracker using OOP.

```python
# Expense dataclass
from dataclasses import dataclass
from datetime import datetime


@dataclass
class Expense:
    date: str
    category: str
    amount: float
    description: str

    @property
    def month(self) -> str:
        return self.date[:7]


e = Expense("2025-06-15", "Food", 12.50, "Lunch")
print(e)
print(f"Month: {e.month}")
```

```python
# ExpenseTracker stub
import csv
from pathlib import Path
from typing import List


class ExpenseTracker:
    """Manages expenses with CSV persistence."""

    def __init__(self, filepath: str = "data/expenses.csv") -> None:
        self.filepath = Path(filepath)
        self.filepath.parent.mkdir(exist_ok=True)
        if not self.filepath.exists():
            with open(self.filepath, "w", newline="", encoding="utf-8") as f:
                writer = csv.writer(f)
                writer.writerow(["date", "category", "amount", "description"])

    def add_expense(self, expense: Expense) -> None:
        with open(self.filepath, "a", newline="", encoding="utf-8") as f:
            writer = csv.writer(f)
            writer.writerow([expense.date, expense.category, expense.amount, expense.description])

    def get_all(self) -> List[Expense]:
        result: List[Expense] = []
        with open(self.filepath, "r", newline="", encoding="utf-8") as f:
            reader = csv.DictReader(f)
            for row in reader:
                result.append(Expense(
                    date=row["date"],
                    category=row["category"],
                    amount=float(row["amount"]),
                    description=row["description"],
                ))
        return result


tracker = ExpenseTracker()
tracker.add_expense(Expense("2025-06-01", "Food", 10.0, "Breakfast"))
tracker.add_expense(Expense("2025-06-02", "Transport", 5.0, "Bus"))
for exp in tracker.get_all():
    print(f"{exp.date} | {exp.category} | ${exp.amount:.2f} | {exp.description}")
```
