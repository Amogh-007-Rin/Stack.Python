# Module 016: Loop Control (break, continue, loop else)

Take control of your loops with `break`, `continue`, and `else`.

## 1. `break` — Exit the Loop

```python
for num in range(1, 101):
    if num == 5:
        break
    print(num, end=" ")
print("\n--- Loop broken ---")
```

```python
# Search with break
numbers = [3, 7, 15, 22, 8, 10]
target = 22
found = False
for num in numbers:
    if num == target:
        found = True
        print(f"Found {target}!")
        break
if not found:
    print(f"{target} not found.")
```

## 2. `continue` — Skip to Next Iteration

```python
print("Odd numbers 1-10:")
for num in range(1, 11):
    if num % 2 == 0:
        continue
    print(num, end=" ")
print()
```

```python
# Extract digits from a string
text = "Hello, 123 World! 456"
digits = ""
for char in text:
    if not char.isdigit():
        continue
    digits = digits + char
print(f"Digits: {digits}")
```

## 3. Loop `else`

```python
# Else runs when no break occurs
for n in range(2, 10):
    if n % 7 == 0:
        print(f"{n} is divisible by 7 — breaking")
        break
else:
    print("No numbers divisible by 7 in range 2-9")
```

```python
# Prime check with for-else
num = 17
for i in range(2, num):
    if num % i == 0:
        print(f"{num} is not prime ({i} x {num // i})")
        break
else:
    print(f"{num} is prime!")
```

## 4. Nested Loop Control

```python
# break only affects the innermost loop
for i in range(1, 4):
    print(f"Outer: i={i}")
    for j in range(1, 4):
        if j == 2:
            break
        print(f"  Inner: j={j}")
```

```python
# Breaking out of nested loops with a flag
found = False
for i in range(5):
    for j in range(5):
        if i * j == 12:
            print(f"Found: {i} x {j} = 12")
            found = True
            break
    if found:
        break
```

## 6. Find Primes Up to 50

```python
for num in range(2, 51):
    for divisor in range(2, int(num ** 0.5) + 1):
        if num % divisor == 0:
            break
    else:
        print(num, end=" ")
print()
```
