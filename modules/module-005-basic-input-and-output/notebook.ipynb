## The `input()` Function

`input()` waits for the user to type something and press Enter. It returns a **string**.

In a notebook, `input()` works but you must run the cell and then type in the text box that appears.

```python
name = input("What is your name? ")
print("Hello,", name)
```

## Input Always Returns a String

Even if you type a number, `input()` gives you a string.

```python
age = input("How old are you? ")
print("Type:", type(age))
```

## Converting Input to Numbers

Use `int()` or `float()` on the result to do math.

```python
age = int(input("How old are you? "))
print("Next year you will be", age + 1)
```

## Multiple Inputs

Ask for several pieces of information and combine them.

```python
name = input("Enter your name: ")
color = input("Enter your favorite color: ")
print(name, "likes", color)
```

## Simple Interactive Calculator

Get two numbers from the user and do arithmetic.

```python
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
print("Sum:", num1 + num2)
print("Product:", num1 * num2)
```

## Simulating Input (for environments without interactive prompt)

In some environments, you can simulate input by assigning a value directly.

```python
simulated_input = "Alice"
print("Simulated input:", simulated_input)
print("Hello,", simulated_input)
```

## Complete Program: Personal Greeting

This is what a complete interactive program looks like (simulated here with pre-set values):

```python
name = "Alice"
age = 30
city = "London"

print("Hello,", name + "!")
print("You are", age, "years old and live in", city + ".")
print("In 5 years, you will be", age + 5, "years old.")
```

## Interest Calculator

A simple financial calculation using user input (simulated).

```python
principal = 1000.0
rate = 5.0
years = 3

interest = principal * (rate / 100) * years
total = principal + interest

print("Principal:", principal)
print("Interest rate:", rate, "%")
print("Years:", years)
print("Interest earned:", interest)
print("Total:", total)
```

## Key Takeaways

- `input()` always returns a **string**.
- Convert numeric input with `int()` or `float()` before calculations.
- Use the prompt parameter to tell the user what to type.
- End prompt strings with a space for readability.
- Interactive programs: input → process → output.
- In notebooks, `input()` works but may show a text widget.
