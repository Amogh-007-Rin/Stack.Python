## Step 1: Define the Game World

The world is a dictionary of rooms, each with a name, description, exits, and items:

```python
WORLD = {
    "entrance": {
        "name": "Dark Entrance",
        "description": "You stand at the entrance of a dark cave. A cold wind blows from within.",
        "exits": {"north": "hallway"},
        "items": ["torch"]
    },
    "hallway": {
        "name": "Stone Hallway",
        "description": "A long hallway lit by flickering torches.",
        "exits": {"south": "entrance", "east": "treasure_room", "west": "storage"},
        "items": []
    },
    "storage": {
        "name": "Storage Room",
        "description": "A dusty storage room filled with old crates.",
        "exits": {"east": "hallway"},
        "items": ["key"]
    },
    "treasure_room": {
        "name": "Treasure Room",
        "description": "A glittering room filled with gold and jewels!",
        "exits": {"west": "hallway"},
        "items": ["treasure"]
    }
}

print(f"World loaded with {len(WORLD)} rooms.")
print(f"Rooms: {', '.join(WORLD.keys())}")
```

## Step 2: Core Functions

Functions for looking up rooms, moving, and taking items:

```python
def get_room(room_id):
    """Get room data by ID.

    Args:
        room_id: String key for the room

    Returns:
        Room dict or None
    """
    return WORLD.get(room_id)

def move_player(current_room, direction):
    """Move player in a direction.

    Args:
        current_room: Current room ID
        direction: Direction to move (north, south, east, west)

    Returns:
        New room ID, or current room ID if blocked
    """
    room = get_room(current_room)
    exits = room["exits"]
    if direction in exits:
        return exits[direction]
    print("You can't go that way.")
    return current_room

def describe_room(room_id):
    """Print the description of a room.

    Args:
        room_id: Room to describe
    """
    room = get_room(room_id)
    print(f"\n=== {room['name']} ===")
    print(room["description"])
    if room["items"]:
        print(f"Items here: {', '.join(room['items'])}")
    exits = list(room["exits"].keys())
    print(f"Exits: {', '.join(exits)}")

# Test
describe_room("entrance")
next_room = move_player("entrance", "north")
describe_room(next_room)
```

## Step 3: Inventory Management

Functions for taking and dropping items:

```python
def take_item(item, room_id, inventory):
    """Take an item from a room and add it to inventory.

    Args:
        item: Name of the item to take
        room_id: Current room ID
        inventory: List of player items

    Returns:
        True if item was taken, False otherwise
    """
    room = get_room(room_id)
    if item in room["items"]:
        room["items"].remove(item)
        inventory.append(item)
        print(f"You take the {item}.")
        return True
    print(f"There is no {item} here.")
    return False

def show_inventory(inventory):
    """Display the player's inventory.

    Args:
        inventory: List of player items
    """
    if not inventory:
        print("You are carrying nothing.")
    else:
        print(f"Inventory: {', '.join(inventory)}")

# Test
inv = []
take_item("torch", "entrance", inv)
show_inventory(inv)
```

## Step 4: The Complete Game Engine

Here's the core game loop that parses commands and updates state:

```python
def parse_command(cmd, current_room, inventory, game_flags):
    """Parse and execute a command.

    Args:
        cmd: Raw input string
        current_room: Current room ID
        inventory: List of player items
        game_flags: Dict of game state flags

    Returns:
        Tuple of (new_room, continue_playing)
    """
    cmd = cmd.strip().lower()
    parts = cmd.split()
    if not parts:
        return current_room, True

    action = parts[0]

    if action in ("quit", "exit"):
        print("Thanks for playing!")
        return current_room, False

    elif action in ("look", "l"):
        describe_room(current_room)

    elif action in ("north", "n"):
        current_room = move_player(current_room, "north")

    elif action in ("south", "s"):
        current_room = move_player(current_room, "south")

    elif action in ("east", "e"):
        current_room = move_player(current_room, "east")

    elif action in ("west", "w"):
        current_room = move_player(current_room, "west")

    elif action == "take":
        if len(parts) < 2:
            print("Take what?")
        else:
            take_item(parts[1], current_room, inventory)

    elif action in ("inventory", "i"):
        show_inventory(inventory)

    elif action == "help":
        print("Commands: look, north/south/east/west, take [item], inventory, help, quit")

    else:
        print(f"Unknown command: {action}")

    return current_room, True

def game_loop():
    """Run the main game loop."""
    current_room = "entrance"
    inventory = []
    game_flags = {}

    print("=== Text Adventure Game ===")
    print("Type 'help' for commands.")
    describe_room(current_room)

    playing = True
    while playing:
        try:
            cmd = input("\n> ")
            current_room, playing = parse_command(cmd, current_room, inventory, game_flags)
        except (EOFError, KeyboardInterrupt):
            print("\nThanks for playing!")
            playing = False

print("Game engine ready. Run game_loop() to play!")
print("\n=== Quick Demo ===")
current_room = "entrance"
inv = []
flags = {}
for cmd in ["look", "take torch", "north", "look", "east", "look", "i", "quit"]:
    print(f"\n> {cmd}")
    current_room, playing = parse_command(cmd, current_room, inv, flags)
    if not playing:
        break
```
