## Positional vs Keyword Arguments

Positional — matched by position; Keyword — matched by name:

```python
def power(base, exp):
    """Raise base to the exponent power.

    Args:
        base: The base number
        exp: The exponent

    Returns:
        base raised to exp
    """
    return base ** exp

print(power(2, 3))        # positional
print(power(exp=3, base=2))  # keyword (order doesn't matter)
```

## Default Parameter Values

Make parameters optional with defaults:

```python
def greet(name, greeting="Hello"):
    """Greet someone with an optional greeting.

    Args:
        name: The person's name
        greeting: The greeting word (default "Hello")
    """
    print(f"{greeting}, {name}!")

greet("Alice")
greet("Bob", "Hi")
greet("Charlie", greeting="Hey")
```

## *args — Variable Positional Arguments

Captures extra positional args as a tuple:

```python
def sum_all(*args):
    """Sum any number of arguments.

    Args:
        *args: Variable number of numeric arguments

    Returns:
        Sum of all arguments
    """
    return sum(args)

print(sum_all(1, 2, 3, 4))
print(sum_all(5, 10))
print(sum_all())
```

## **kwargs — Variable Keyword Arguments

Captures extra keyword args as a dict:

```python
def print_info(**kwargs):
    """Print keyword arguments as key-value pairs.

    Args:
        **kwargs: Arbitrary keyword arguments
    """
    for key, value in kwargs.items():
        print(f"{key}: {value}")

print_info(name="Alice", age=30, city="New York")
```

## Argument Ordering

Order: positional → default → *args → **kwargs:

```python
def example(pos, default_val=5, *args, **kwargs):
    """Demonstrate argument ordering.

    Args:
        pos: A positional parameter
        default_val: A parameter with default (5)
        *args: Extra positional args
        **kwargs: Extra keyword args
    """
    print(f"pos={pos}, default={default_val}")
    print(f"args={args}")
    print(f"kwargs={kwargs}")

example(1, 2, 3, 4, 5, x=10, y=20)
```

## Unpacking with * and ** When Calling

Unpack iterables with `*` and dicts with `**`:

```python
def add(a, b, c):
    """Return sum of three numbers."""
    return a + b + c

nums = [1, 2, 3]
print(add(*nums))  # 6
```

```python
def create_user(name, age, email):
    """Print user info."""
    print(f"{name}, {age}, {email}")

data = {"name": "Alice", "age": 30, "email": "alice@example.com"}
create_user(**data)
```

## Key Takeaways

- Positional → by position; Keyword → by name.
- Default parameters make arguments optional.
- `*args` captures extra positional args as tuple.
- `**kwargs` captures extra keyword args as dict.
- Order: positional → default → *args → **kwargs.
- Use `*`/`**` to unpack sequences/dicts when calling.
